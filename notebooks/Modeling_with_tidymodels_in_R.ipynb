#####**5. Formules types**

+ Avant que le panais puisse adapter un modèle aux données, il nécessite que les colonnes soient affectées à un rôle de résultat ou de prédicteur. 

+ Ceci est fait avec des formules R et suit la forme générale de la variable de résultat à gauche suivie d’un tilde, puis d’une ou plusieurs variables prédictives séparées par des signes plus. 

+ Pour utiliser toutes les colonnes disponibles dans une trame de données comme prédicteurs, la notation abrégée du point tilde de résultat peut être utilisée. Pour prédire hwy en utilisant cty, nous utiliserions hwy tilde cty dans notre formule de modèle.

#####**6. Le paquet panais**

+ Le paquet panais fournit une syntaxe unifiée pour la spécification du modèle dans R. 

+ La création d’un objet de modèle de panais implique de spécifier le type de modèle, tel que la régression linéaire, le moteur de calcul, qui spécifie le package sous-jacent qui sera utilisé pour ajuster le modèle, et le mode qui est la régression ou la classification. 

+ Avec le panais, il est possible d’ajuster une régression linéaire avec le moteur traditionnel 'lm' fourni en base R ou le moteur 'stan', qui estime les paramètres du modèle en utilisant l’estimation bayésienne des paramètres. 

+ La puissance du panais réside dans le fait qu’il combine un grand nombre de packages d’apprentissage automatique qui correspondent au même type de modèle en utilisant une syntaxe commune.

####**7. Ajustement d’un modèle de régression linéaire**


+ Pour adapter notre modèle de régression linéaire, nous commençons par définir un objet modèle de panais nommé lm_model. 

+ Nous utilisons **la fonction linear_reg()** pour créer un objet de modèle de régression linéaire. 

+ Ensuite, nous le passons dans la fonction **set_engine** où nous spécifions le moteur 'lm'. 

+ Enfin, nous passons les résultats dans **la fonction set_mode()** où nous spécifions « régression » puisque nous prédisons une variable de résultat numérique. 

+ Pour entraîner notre modèle, nous passons lm_model à **la fonction fit()** et fournissons une formule de modèle et les données sur lesquelles entraîner le modèle.

####**8. Obtention des paramètres estimés**

####**9. Faire des prédictions**


+ Les prédictions du modèle sont obtenues en passant le modèle entraîné, lm_fit, à **la fonction predict()**.

+ La ligne $y = x$ est également tracée et représente le cas où toutes les prédictions et les valeurs de résultat sont égales, ce qui donne une valeur R au carré de 1. 

+ Les diagrammes R carrés sont utiles pour identifier les problèmes liés à la performance du modèle, tels que les relations non linéaires entre la variable de résultat et les prédicteurs ou les régions où le modèle peut être systématiquement sous-prédictif ou surestimé.

####**2. Classification binaire**

+ ***En classification binaire, nous avons deux catégories possibles***. 

+ La classe positive est la catégorie qu’il est intéressant de prédire, comme « oui » dans notre variable achetée. 

+ La classe négative est la catégorie restante. 

+ Dans les modèles ordonnés, la variable de résultat doit être un facteur avec le premier niveau comme classe positive. 

+ Pour vérifier l’ordre d’un vecteur factoriel, passez-le dans **la fonction levels()**. 

+ Pour modifier l’ordre, affectez un vecteur de caractère avec l’ordre souhaité à cet appel de **fonction levels()**.

####**3. Matrice de confusion**

+ ***Une matrice de confusion affiche le nombre de toutes les combinaisons de résultats réels et prévus à partir d’un modèle de classification binaire***. 

+ Les catégories prédites apparaissent le long des lignes tandis que les résultats réels apparaissent le long des colonnes. 

+ Il existe deux types de prédictions correctes, connues sous le nom de vrais positifs et de vrais négatifs le long de la diagonale. 

+ Un modèle de classification binaire peut générer deux types d’erreurs. 

+ Prédiction d’une classe positive pour une ligne qui a une classe négative réelle, appelée faux positif, et prédiction d’une classe négative pour une ligne avec une classe positive réelle, appelée faux négatif.

####**4. Mesures de classification à l’aune**

+ La précision n’est généralement pas la meilleure mesure.

+ ***La sensibilité est la proportion de tous les cas positifs qui ont été correctement classés et s’améliore avec des taux de faux négatifs plus faibles***. 

+ Par exemple, parmi les clients qui ont acheté un produit, quelle proportion notre modèle prédit-il correctement?

####**9. Spécificité**

+ Par exemple, parmi les clients qui n’ont pas acheté un produit, quelle proportion notre modèle prédit-il correctement?

####**2. Ingénierie des fonctionnalités avec le package de recettes**

+ L’ingénierie des fonctionnalités est réalisée avec le package de recettes. 

+ Il est conçu pour aider à toutes les étapes de l’ingénierie des fonctionnalités, notamment l’attribution de rôles variables aux colonnes de nos données, la définition des tâches de prétraitement et des transformations de données, la formation de nos transformations de données et leur application à de nouvelles sources de données.

####**3. Spécification des types de variables et des rôles**


+ La première étape de l’ingénierie des fonctionnalités consiste à attribuer chaque colonne de nos données à un rôle de résultat ou de prédicteur et à déterminer leur type de données, qui peut être numérique ou catégorique. 

+ **La fonction recipes()** est utilisée pour cette tâche.

####**4. Étapes de prétraitement des données**

+ L’étape suivante consiste à définir une séquence d’étapes de **prétraitement des données**, qui peuvent inclure **l’imputation des données manquantes**, **le centrage et la mise à l’échelle des variables numériques, la création de nouvelles variables à partir de ratios de variables existantes et bien d’autres possibilités**. 

+ Ces transformations sont codées avec des fonctions ***uniques step_*()***.

  + https://recipes.tidymodels.org/reference/index.html

####**5. Étapes de prétraitement de la formation**

+ Une fois les étapes de prétraitement définies, elles doivent être formées et estimées à l’aide de données. 

+ Cela inclut des éléments tels que ***le calcul de la moyenne et de l’écart-type des colonnes numériques pour le centrage et la mise à l’échelle des données et le stockage de formules pour la création de nouvelles colonnes****. 

+ **La fonction prep()** est utilisée pour cette tâche.

####**6. Application de recettes à de nouvelles données**

+ La dernière étape de l’ingénierie des fonctionnalités consiste à appliquer les transformations de données entraînées aux jeux de données d’apprentissage et de test, ainsi qu’aux nouvelles sources de données pour les prévisions futures. 

+ Il s’agit d’une étape importante, car les algorithmes d’apprentissage automatique nécessitent le même format de données que celui utilisé lors de la formation du modèle pour prédire de nouvelles valeurs. 

+ **La fonction bake()** des recettes est utilisée pour cette tâche.

+ Les données nominales doivent être transformées en données numériques lors de l’ingénierie des fonctionnalités, car de nombreux algorithmes d’apprentissage automatique nécessitent une saisie numérique. 

+ **Le codage One-Hot est une transformation qui mappe les valeurs distinctes d’une variable nominale à une séquence de variables indicateurs 0/1**.

+ Si nous avons n valeurs distinctes dans nos données catégorielles, nous obtiendrons $n - 1$ variables indicatrices. 

+ Il s’agit de la méthode préférée pour les tidymodels et de la méthode par défaut dans le package de recettes. 

+ Avec cette méthode, notre variable de département est mappée à une séquence de deux variables indicatrices.

###**Estimation des performances avec validation croisée**

####**1. Estimation des performances avec validation croisée**

+ Dans cette section, nous allons apprendre comment améliorer le processus d’évaluation du modèle avec une méthode connue sous le nom de validation croisée.

####**2. Ensembles de données de formation et de test**

+ Nous avons créé des ensembles de données de formation et de test dans notre processus de modélisation où les données d’apprentissage sont utilisées pour l’ajustement du modèle tandis que les données de test sont réservées à l’évaluation du modèle pour se prémunir contre le surajustement. 
+ Un inconvénient de cette méthode est que nous n’obtenons qu’une seule estimation des performances du modèle.

####**3. Validation croisée K-fold**

+ La validation croisée K-fold est une technique qui fournit des estimations K des performances du modèle et est généralement utilisée pour comparer différents types de modèles, tels que la régression logistique et les arbres de décision.

####**4. Validation croisée K-fold**

+ Les données d’entraînement sont partitionnées aléatoirement en K ensembles de taille à peu près égale, appelés plis, qui sont utilisés pour effectuer K itérations d’ajustement et d’évaluation du modèle. 

+ Le jeu de données de test est exclu de ce processus afin qu’il puisse fournir une estimation finale et indépendante des performances du modèle une fois qu’un type de modèle est choisi.

####**5. Apprentissage automatique avec validation croisée**

+ Si nous avons 5 volets, nous aurons cinq itérations de formation et d’évaluation du modèle.

####**6. Apprentissage automatique avec validation croisée**

+ Dans la première itération, le pli 1 est réservé à l’évaluation du modèle tandis que les autres sont utilisés pour l’apprentissage du modèle.

####**7. Apprentissage automatique avec validation croisée**

+ Dans la deuxième itération, le pli 2 est réservé à l’évaluation du modèle tandis que les autres sont utilisés pour l’apprentissage du modèle.

####**8. Apprentissage automatique avec validation croisée**

+ Ce processus se poursuit jusqu’à la cinquième itération, où le pli 5 est utilisé pour l’évaluation du modèle. 

+ Au total, cela fournit cinq estimations de la performance du modèle.

####**9. Création de plis de validation croisée**

+ **La fonction vfold_cv()** crée des plis de validation croisée et prend un tibble comme premier argument, un nombre de plis comme second et une variable de stratification comme troisième. 

+ Pour créer 10 plis à partir de nos données leads_training, nous définissons v égal à 10 et stratifions par achat pour nous assurer que chaque pli a des proportions similaires des valeurs de résultat. 

+ L’exécution de **la fonction set-dot-seed()** avant **la fonction vfold_cv()** garantit la reproductibilité. 

+ Cette fonction prend n’importe quel entier comme argument et définit la graine du générateur de nombres aléatoires de R. 

+ Il en résulte un tibble avec une colonne de liste nommée splits et une colonne id qui identifie chaque pli. 

+ Chaque ligne des fractionnements contient un objet de fractionnement de données qui contient les instructions pour fractionner le pli de cette ligne en un jeu d’apprentissage ou d’évaluation.

####**10. Formation sur modèle avec validation croisée**

+ **La fonction fit_resamples()** effectue une validation croisée dans les tidymodels. 

+ Pour entraîner nos leads_workflow sur chaque pli, nous les passons à **fit_resamples()**, fournissons **leads_folds** à l’argument rééchantillons et notre fonction métrique personnalisée à l’argument métrique optionnel. 

+ Par défaut, la précision et l’AUC ROC sont calculées. 

+ Cela renvoie un objet rééchantillonné sur lequel nous pouvons collecter des métriques. 

+ Nous voyons que chaque métrique a été estimée 10 fois, une pour chaque pli. 

+ La moyenne de ces estimations est fournie dans la colonne moyenne.

####**11. Résultats détaillés de la validation croisée**


+ Passer résume est égal à faux dans **collect_metrics()** créera un tibble avec des résultats détaillés. 

+ Pour notre leads_rs_fit, cela nous donne 30 lignes au total ce qui représente nos 3 métriques multipliées par nos 10 plis. 

+ La colonne dot_metric identifie la métrique tandis que la colonne d’estimation ponctuelle fournit la valeur estimée pour chaque pli.

####**12. Résumer les résultats de la validation croisée**

+ Les résultats de **collect_metrics()** peuvent être résumés par dplyr. 

+ En commençant par rs_metrics, nous regroupons par colonne point-métrique, puis calculons les statistiques récapitulatives avec la fonction **summarize()** pour chaque métrique de la colonne point-métrique.

####**13. Méthodologie de validation croisée**

####**Réglage des hyperparamètres**

####**1. Réglage des hyperparamètres**

+ ****Dans cette section, nous nous concentrerons sur le réglage des hyperparamètres, qui est une autre méthode d’optimisation des performances du modèle.***

####**2. Hyperparamètres**


+ ***Les hyperparamètres sont des paramètres de modèle dont les valeurs sont définies avant l’apprentissage du modèle et contrôlent la complexité du modèle.*** 

+ Dans le panais, les arbres de décision ont trois hyperparamètres. 

+ cost_complexity est un nombre qui est utilisé pour pénaliser les arbres avec un grand nombre de nœuds terminaux. 

+ **tree_depth** contrôle la longueur du chemin d’accès de la racine à n’importe quel nœud terminal. 
+ **min_n** contrôle les points de données minimaux requis dans un nœud pour un fractionnement ultérieur.

####**3. Valeurs d’hyperparamètre par défaut**

+ Lorsqu’aucun argument n’est fourni à ***la fonction decision_tree(), cost_complexity est défini sur 0-point-01, tree_depth à 30 et min_n à 20***. 

+ Toutefois, ces valeurs peuvent ne pas être optimales pour tous les jeux de données. 
+ Les changer pourrait améliorer les performances globales. 

+ ****Le réglage des hyperparamètres est le processus d’utilisation de la validation croisée pour trouver l’ensemble optimal de valeurs d’hyperparamètres pour un modèle.***

####**4. Étiquetage des paramètres hymétriques pour le réglage**


+ **La fonction tune()** du package tune est utilisée pour étiqueter les hyperparamètres du modèle.
+  Nous définissons simplement les valeurs de chaque hyperparamètre sur **tune()** dans **decision_tree()**. 

+ Cela crée un objet modèle avec des paramètres de réglage et indiquera aux autres méthodes qu’elles doivent être optimisées. 

+ Lorsque nous imprimons notre objet dt_tune_model, les hyperparamètres apparaissent sous les arguments principaux.

####**5. Création d’un flux de travail de réglage**

+ Pour effectuer le réglage des hyperparamètres, nous créons d’abord un nouveau flux de travail en mettant à jour notre précédent avec notre nouveau modèle d’arbre de décision. 

+ Nous passons **leads_wkfl** à **update_model()** et remplaçons notre arbre de décision précédent par notre nouveau. 

+ Lorsque nous imprimons notre nouveau flux de travail, nous voyons que les hyperparamètres apparaissent dans la spécification du modèle avec notre recette d’ingénierie des fonctionnalités.

####**6. Recherche de grille**


+ Le réglage des hyperparamètres est effectué à l’aide de la recherche par grille, une méthode dans laquelle une grille de valeurs d’hyperparamètres est générée. 

+ Pour chaque combinaison, la validation croisée est utilisée pour estimer les performances du modèle et la combinaison la plus performante est choisie.

####**7. Identification des hyperparamètres**

+ La fonction parameters du package de cadrans peut être utilisée pour identifier les hyperparamètres dans un objet modèle de panais. 

+ Cette fonction renvoie un tibble avec les hyperparamètres étiquetés par **tune()**, le cas échéant. 

+ Ce tibble est utilisé pour aider à générer des grilles pour le réglage.

####**8. Grille aléatoire**


+ Une méthode populaire pour la recherche de grille consiste à générer des combinaisons aléatoires de valeurs d’hyperparamètres. 

+ Comme il peut y avoir un nombre infini de combinaisons d’hyperparamètres, les choisir au hasard donnera une plus grande chance de découvrir des combinaisons optimales. 

+ Les choisir de manière systématique peut limiter la gamme de valeurs choisies, ce qui pourrait conduire à de mauvais résultats. 

+ Pour créer une grille aléatoire, nous passons les résultats de la fonction **parameters()** sur notre **dt_tune_model** à **la fonction grid_random()** et sélectionnons le nombre de combinaisons à générer. 

+ Pour reproduire cette grille, exécutez ***la fonction set-dot-seed() avant grid_random()***. 
+ Cela crée un tibble avec des combinaisons aléatoires.

####**9. Enregistrement d’une grille de réglage**

+ Pour effectuer un réglage des hyperparamètres sur notre modèle d’arbre de décision, nous enregistrons d’abord les résultats de ***grid_random()***, où nous avons 5 combinaisons.

####**10. Réglage des hyperparamètres avec validation croisée**

+ Nous passons ensuite notre leads_tune_wkfl à **la fonction tune_grid()**. 

+ Cette fonction nécessite également un objet de validation croisée, une grille de réglage et une fonction de métriques personnalisée facultative. 

+ Pour notre flux de travail, nous utilisons leads_folds, notre dt_grid et notre fonction de leads_metrics personnalisée. 

+ **tune_grid()** renvoie une batterie de résultats de réglage. 

+ La colonne dot-metrics est une colonne de liste avec les résultats pour chaque pli.

####**11. Exploration des résultats de réglage**

+ ***Les fonctions collect_metrics() peuvent être utilisées sur un objet de réglage pour collecter des résultats.*** 

+ Par défaut, la mesure de performances moyennes est renvoyée pour chaque combinaison de valeurs d’hyperparamètres et de mesures de performances.

Puisque nous avons 10 plis dans notre objet de validation croisée, chaque ligne est la moyenne des performances sur les 10 plis.
