####**3. Distance vs similitude**
+ **Vous voyez, la plupart des méthodes de regroupement mesurent la similitude entre les observations à l’aide d’une métrique de dissmilarité, souvent appelée distance**. 

+ Ces deux concepts ne sont que les deux faces d’une même médaille. 

+ Si deux observations ont une grande distance, elles sont **moins semblables l’une à l’autre**. 

+ De même, si leur valeur de distance est petite, **alors ils sont plus similaires**. 

+ Naturellement, nous devrions d’abord développer une intuition aiguë par ce que l’on entend par distance.

+ Jusqu'à présent, vous n'avez travaillé qu'avec un seul type de mesure de distance, **la distance euclidienne**. 

+ Cette métrique est couramment utilisée et constitue un excellent point de départ lorsque vous travaillez avec des données continues. 

+ **Mais que se passe-t-il si les données que vous avez ne sont pas continues mais catégoriques ?**

####**2. Les données binaires**
+ Commençons par le cas le plus élémentaire **des caractéristiques catégorielles, celles qui sont binaires, ce qui signifie que les valeurs ne peuvent être que l'une des deux possibilités**. D

+ ans le cas présent, il s'agit de données d'enquête, appelons-la enquête a. 

+ On a demandé aux participants à cette enquête s'ils aiment boire différents types de boissons alcoolisées. 

+ Comme ils ne peuvent répondre que par **oui ou non**, nous pouvons coder cette **réponse binaire comme VRAI ou FAUX**. 

+ Nous aimerions savoir quels participants sont similaires les uns aux autres sur la base de leurs réponses. 

+ Pour le calculer, nous utiliserons **le score de similarité appelé indice de Jaccard**.

####**3. L'indice de Jaccard**


+ **Cette mesure de similarité représente le rapport entre l'intersection de A et B et l'union de A et B**. 

+ Ou, plus intuitivement, **le rapport entre le nombre de fois où les caractéristiques des deux observations sont VRAIES et le nombre de fois où elles sont toujours VRAIES**. 

+ Revenons donc à l'exemple précédent.

####**4. Calcul de la distance de Jaccard**

+ Calculons la similarité de Jaccard pour deux observations un et deux. 

+ Elles ne se rejoignent que dans une seule catégorie, la bière, donc pour l'intersection, nous obtenons la valeur un. 

+ Alors que le nombre de catégories dans lesquelles ces observations sont toujours vraies, ou l'union, est de quatre. 

+ En divisant l'intersection par l'union, on obtient la valeur de similarité de Jaccard de 0-point-25. 

+ Mais qu'en est-il de la distance ? 

+ Rappelez-vous que la distance est égale à 1 - similarité, donc dans ce cas la distance est juste de 0-point-75.

###**Introduction aux K-means**

####**1. Introduction aux K-means**

+ Dans le dernier chapitre, **vous avez appris à utiliser la méthode de clustering hiérarchique pour regrouper les observations**. 

+ Dans ce chapitre, **vous découvrirez une autre méthode populaire de clustering appelée clustering k-means**.

###**2. K-means**

+ Ces données se composent de douze joueurs sur un terrain de football au début du match. 

+ À ce stade du jeu, les équipes sont positionnées sur des côtés opposés du terrain. 

+ Nous nous attendons à ce que le regroupement puisse être efficace pour identifier les équipes et affecter chaque joueur à la bonne équipe. 

+ **La première étape du clustering k-means consiste à prendre une décision sur le nombre de clusters à générer**. 

+ Il s’agit du clustering k dans k-means. 

+ Cela peut être décidé à l’avance en fonction de notre compréhension des données ou il peut être estimé à partir des données empiriquement. 

+ Nous discuterons des méthodes d’estimation plus loin dans ce chapitre. 

+ Dans cet exemple, nous pouvons tirer parti de ce que l’on sait de nos données. 

+ Puisque nous savons que le football se joue avec deux équipes, nous pouvons utiliser un k de 2 pour le nombre de clusters souhaité. Une fois k établi, l’algorithme peut continuer.

####**3. K-means**

+ **La première étape de l’algorithme k-means consiste à initialiser k points à des positions aléatoires dans l’espace des entités, nous appellerons ces points les centroïdes de cluster**. 

+ Dans ces données, nous illustrerons nos deux centroïdes à l’aide d’un $x$ rouge et d’un $x$ bleu.

####**4. K-means**

+ Pour chaque observation, **la distance est calculée entre l’observation et chaque centroïde**. 

+ Dans le regroupement des k-means, la distance est euclidienne limitée seulement.

####**5. K-means**

+ Les observations sont initialement assignées **au centroïde dont elles sont les plus proches**.

####**6. K-means**
+ Nous pouvons voir cette limite de décision représentée par l’espace colorimétrique.

####**7. K-means**
+ Les observations ont maintenant une affectation initiale à l’un des deux groupes.

####**8. K-means**
+ L’étape suivante consiste à déplacer les centroïdes vers les points centraux des clusters résultants.

####**9. K-means**
+ Encore une fois, la distance de chaque observation est calculée à chaque centroïde.

####**10. K-means**
+ Et ils sont réaffectés en fonction du centroïde dont ils sont les plus proches.

####**11. K-means**
+ **Ce processus se poursuit jusqu’à ce que les centroïdes se stabilisent et que les observations ne soient plus réassignées**. 

+ **C’est l’algorithme fondamental du clustering kmeans**.

####**12. kmeans()**


+ Pour générer le modèle kmeans dans R, vous utiliserez la fonction du même nom.

+ Centres dans ce cas est synonyme de k.

####**Instructions**

+ Calculez la distance euclidienne entre les professions et stockez-la dans dist_oes.

```r
oes <- readRDS("/content/oes.rds")
head(oes, 3)
```

```r
# Calculate Euclidean distance between the occupations
dist_oes <- dist(oes, method = "euclidean")

# Generate an average linkage analysis 
hc_oes <- hclust(dist_oes, method = "average")

# Create a dendrogram object from the hclust variable
dend_oes <- as.dendrogram(hc_oes)

# Plot the dendrogram
plot(dend_oes)
```

```r

# Color branches by cluster formed from the cut at a height of 100000
dend_colored <- color_branches(dend_oes, h = 100000)

# Plot the colored dendrogram
plot(dend_colored)
```

####**Instructions**

+ Créez le cadre de données df_oes à partir de la matrice de données oes, en vous assurant de stocker le rowname comme une colonne (utilisez **rownames_to_column()** de la bibliothèque tibble).


+ Construisez le vecteur d'affectation des clusters cut_oes en utilisant **cutree()** avec un **h = 100,000**.

```r
dist_oes <- dist(oes, method = 'euclidean')
hc_oes <- hclust(dist_oes, method = 'average')
```

```r
head(dist_oes,3)
```

```r
head(hc_oes)
```

```r
library(tibble)
library(tidyr)

# Use rownames_to_column to move the rownames into a column of the data frame
df_oes <- rownames_to_column(as.data.frame(oes), var = 'occupation')
head(df_oes, 3)
```

```r
# Create a cluster assignment vector at h = 100,000
cut_oes <- cutree(hc_oes, h = 100000)
data.frame(cut_oes)
```

```r
# Generate the segmented the oes data frame
clust_oes <- mutate(df_oes, cluster = cut_oes)
head(clust_oes, 3)
```

```r
# Create a tidy data frame by gathering the year and values into two columns
gathered_oes <- gather(data = clust_oes, 
                       key = year, 
                       value = mean_salary, 
                       -occupation, -cluster)

head(gathered_oes, 3)
```

####**K-means : Analyse du coude**

+ Dans les exercices précédents, vous avez utilisé le dendrogramme pour proposer un clustering qui génère 3 arbres. 

+ Dans cet exercice, vous allez exploiter le diagramme coudé des k-means pour proposer le "meilleur" nombre de clusters.

####**Instructions**

+ Utilisez **map_dbl()** pour exécuter **kmeans()** en utilisant **les données oes** pour des valeurs $k$ allant de **1 à 10** et extraire la somme totale des carrés à l'intérieur des clusters de chaque modèle : 

    + model$tot.withinss. 
    
+ Stockez le vecteur résultant comme **tot_withinss**.

+ Créez un nouveau cadre de données **elbow_df** contenant les valeurs de $k$ et le vecteur de la somme totale des carrés à l'intérieur du cluster.


+ Utilisez les valeurs de **elbow_df** pour tracer un graphique linéaire montrant la relation entre $k$ et la somme totale des carrés à l'intérieur des clusters.



```r
# Use map_dbl to run many models with varying value of k (centers)
tot_withinss <- map_dbl(1:10,  function(k){
  model <- kmeans(x = oes, centers = k)
  model$tot.withinss
})

# Generate a data frame containing both k and tot_withinss
elbow_df <- data.frame(
  k = 1:10,
  tot_withinss = tot_withinss
)

elbow_df
```

```r
# Plot the elbow plot
ggplot(elbow_df, aes(x = k, y = tot_withinss)) +
  geom_line() +
  scale_x_continuous(breaks = 1:10)
```

####**K-means : Largeur moyenne des silhouettes**

+ Ainsi, le clustering hiérarchique aboutit à 3 clusters et la méthode du coude en suggère 2. 

+ Dans cet exercice, utilisez les largeurs moyennes de silhouette pour explorer quelle devrait être la "meilleure" valeur de $k$.

####**Instructions**

+ Utilisez **map_dbl()** pour exécuter **pam()** en utilisant les données oes pour des valeurs k allant de 2 à 10 et extraire la valeur de la largeur moyenne de la silhouette de chaque modèle : 
      model$silinfo$avg.width. 
  
+ Stockez le vecteur résultant sous le nom de sil_width.


+ Construisez un nouveau cadre de données sil_df contenant les valeurs de $k$ et le vecteur des largeurs moyennes de silhouette.


+ Utilisez les valeurs de sil_df pour tracer un graphique linéaire montrant la relation entre k et la largeur moyenne de la silhouette.

```r
# Use map_dbl to run many models with varying value of k
sil_width <- map_dbl(2:10,  function(k){
  model <- pam(oes, k = k)
  model$silinfo$avg.width
})

# Generate a data frame containing both k and sil_width
sil_df <- data.frame(
  k = 2:10,
  sil_width = sil_width
)
sil_df
```

```r
# Plot the relationship between k and sil_width
ggplot(sil_df, aes(x = k, y = sil_width)) +
  geom_line() +
  scale_x_continuous(breaks = 2:10)
```
