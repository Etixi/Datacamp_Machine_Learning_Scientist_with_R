####**4. Avantages des modèles arborescents**

+ L’un des plus grands avantages des arbres de décision est qu’ils sont faciles à expliquer.

+ **Les valeurs manquantes ne sont pas un problème et les arbres sont robustes aux valeurs aberrantes**. 

+ Enfin, il est relativement **rapide** d’entraîner un arbre de décision, de sorte que les méthodes arborescentes peuvent **gérer de grands ensembles de données**.

####**4. Répartition des données**

+ **Une solution courante à ce problème consiste à diviser les données en deux groupes**. 

+ **Une partie des données, l’ensemble d’apprentissage, est utilisée pour estimer les paramètres et comparer ou ajuster les modèles**. 

+ **L’ensemble de test est gardé en réserve jusqu’à la fin du projet**. 

+ Il est utilisé comme source impartiale pour ***mesurer la performance finale du modèle***. 

+ Il est essentiel de ne pas utiliser l’ensemble de test avant ce point, sinon les données de test feront partie du processus de développement du modèle.

####**5. Méthodes de fractionnement**
+ Il existe différentes façons de créer ces partitions de données. 

+ Vous pouvez prendre des données de la fin, du centre ou d’un échantillon aléatoire. 

+ Il suffit de sélectionner une partie de l’ensemble de données, disons 80% de tous les échantillons, pour l’ensemble d’apprentissage et d’utiliser le reste pour l’ensemble de test. 

+ Nous présenterons des méthodes de fractionnement plus intéressantes et plus robustes au chapitre 2.

####**6. La fonction initial_split()**

+ **La fonction initial_split() du paquet rsample** est pratique car elle fait tout le travail pour vous. 

+ La fonction prend les données d’origine, le diabète et un argument de proportion, disons 0 point-9, et attribue au hasard des échantillons (lignes de la tibble) à l’ensemble d’entraînement ou de test. 

+ Si aucune proportion n’est donnée, il attribue 75% des échantillons à l’entraînement et 25% à l’ensemble de test. 

+ Le résultat est un objet divisé qui contient des ensembles d’analyse et d’évaluation, qui ne sont que des noms différents pour les ensembles d’apprentissage et de test. 

+ Ici, nous voyons que 76 échantillons sont dans l’ensemble de test ou d’évaluation, ce qui représente environ 10% des 768 échantillons au total.

+ **La matrice de confusion est un outil très utile pour évaluer les modèles de classification binaire**. 

+ **C’est ce qu’on appelle une matrice de confusion parce qu’elle révèle à quel point le modèle est confus entre les deux classes et met en évidence les cas dans lesquels il confond une classe pour l’autre**.

####**5. Matrice de confusion**
+ Les colonnes d’une matrice de confusion correspondent aux étiquettes de vérité,

####**6. Matrice de confusion**
+ et les lignes représentent les prédictions.

####**7. Matrice de confusion**
+ Dans un problème de classification binaire, la matrice de confusion sera une table 2 par 2. 

+ La diagonale principale contient le nombre d’exemples correctement classés, c’est-à-dire les prédictions « oui » qui sont en fait « oui », et les prédictions « non » qui sont en fait « non ». 

+ Un bon modèle contiendra la plupart des exemples dans la diagonale principale (les carrés verts) et il aura un petit nombre d’exemples, idéalement zéro, dans la diagonale hors diagonale (les carrés rouges).

###**Mesures de performance pour les arbres de régression**

####**1. Mesures de performance pour les arbres de régression**

+ Bien joué! Maintenant que vous avez formé un arbre de régression, il est temps d’évaluer les performances du modèle.

####**2. Comment mesurer la performance?**

+ En classification, une seule prédiction peut être bonne ou fausse, et rien d’autre. 

+ Il est donc logique d’évaluer les modèles de classification à l’aide d’une métrique telle que la précision ou la matrice de confusion. 

+ En régression, cependant, votre prédiction peut être presque juste ou même totalement fausse. 

+ Il n’y a pas d’exactitude binaire qui calcule la précision. 

+ Vous devez évaluer les arbres de régression à l’aide d’une métrique différente. 

+ Puisque la réponse est un nombre à valeur réelle, et que notre prédiction est la même, il est logique de mesurer à quelle distance nos prédictions sont éloignées des valeurs réelles.

####**3. Mesures communes pour la régression**

+ Il existe plusieurs mesures de régression et deux plus populaires sont ***l’erreur absolue moyenne, ou MAE, et l’erreur quadratique moyenne, également connue sous le nom de RMSE***. 

+ ***MAE est la distance absolue moyenne entre les valeurs réelles (ou observées) et les valeurs prédites***. 

+ Dans cette image, les prédictions sont représentées par la ligne bleue et les vraies valeurs sont les points noirs. 

+ Les barres rouges sont les erreurs de prédiction, et l’erreur absolue moyenne serait la longueur moyenne des barres rouges. 

+ ***L’erreur quadratique moyenne est la racine carrée de la longueur quadratique moyenne des barres rouges.***

####**4. Formules et intuition**

+ La formule pour MAE est très simple. 

+ Vous additionnez toutes les différences absolues entre les valeurs réelles et prédites et les divisez par le nombre de prédictions faites pour obtenir l’erreur moyenne. 

+ Le RMSE est similaire, mais au lieu de prendre la différence absolue entre les erreurs, vous mettez au carré les différences des erreurs. Ensuite, vous ferez la moyenne de ces valeurs

####**5. Formules et intuition**

+ et prenez la racine carrée de l’ensemble. Prendre la racine carrée ramène la métrique à l’échelle d’origine de la réponse. 

+ MAE et RMSE expriment tous deux des erreurs moyennes de prédiction du modèle en unités de la variable d’intérêt. 

+ La principale différence entre les deux est que RMSE punit les grandes erreurs plus sévèrement que MAE. 

+ Étant donné que les erreurs sont au carré avant d’être moyennées, le RMSE accorde un poids relativement élevé aux erreurs importantes. 

+ Cela signifie que le RMSE devrait être plus utile lorsque des erreurs importantes sont particulièrement indésirables.

####**6. Codage : prédictions**

####**Plus grosse erreur, plus grosse sanction**

+ Toutes les erreurs sont mauvaises, mais toutes ne le sont pas également. 

+ Parfois, de grosses erreurs de prédiction sont disproportionnellement plus nuisibles que de petites erreurs.

+ Plus les erreurs sont grosses, plus la sanction est grande : 
  + c'est l'une des caractéristiques de l'erreur quadratique moyenne ou RMSE. 
  + Elle élève au carré les grosses erreurs, ce qui punit ces aberrations plus sévèrement que les petites erreurs.

+ La RMSE peut être calculée à l'aide de la formule suivante, où le ième écart quadratique est le carré de la ième erreur.

 $$
 RMSE = \sqrt{\frac{1}{n} \cdot \sum_{i=1} ^n i\text{th squared_diff}}
 $$

###**Compromis biais-variance**

####**1. Compromis biais-variance**
+ Maintenant que nous avons jeté les bases de la modélisation, vous allez utiliser vos nouvelles compétences pour comprendre un concept très important de l’apprentissage automatique.

####**2. Hyperparamètres**

+ Lors de la création d’un modèle, le modeleur - c’est vous! 
  - ***choisit les paramètres qui définissent le processus d’apprentissage, les hyperparamètres***. 
  
  + ***Un exemple est la profondeur de l’arbre, qui contrôle le nombre de scissions effectuées jusqu’à ce qu’une décision soit prise***. 
  
  + Les détails se trouvent dans la documentation, comme vous le voyez ici. 
  
  + Restez à l’écoute pour le chapitre trois, où nous plongeons plus profondément dans ces hyperparamètres!

####**3. Impact sur la complexité du modèle**

+ À l’aide d’hyperparamètres, vous contrôlez la simplicité ou la complexité de la structure de votre modèle, comme le montrent ces deux exemples. 

+ Nous adaptons un modèle qui construit 2 niveaux de la racine aux nœuds feuilles (c’est ce que tree_depth signifie), et un avec 15 niveaux à nos données d’entraînement. L’impact sur le modèle final est énorme.

####**4. Modèle complexe - surajustement - variance élevée**

+ Imaginez que vous souhaitiez prédire la final_grade dans votre jeu de données sur le chocolat. 

+ Vous construisez un arbre de décision très complexe et constatez qu’il correspond étonnamment bien à vos données d’entraînement. 

+ Dans cet exemple, l’erreur absolue moyenne, calculée par la **fonction mae()**, n’est que d’environ 0 point-2. 

+ Cependant, lorsque vous vérifiez votre modèle sur votre ensemble de test, vous observez de très grandes erreurs. 

+ Dans ce cas, nous disons que votre modèle sur-ajuste les données. 
+ Nous appelons cet effet « variance élevée ». 

+ Notre arbre de décision très sophistiqué a très bien appris la structure des données d’entraînement et ne peut pas s’adapter à la structure différente de l’ensemble de test.

####**5. Modèle simple - sous-ajustement - biais élevé**

+ Il est facile de simplifier votre arbre de décision, par exemple en utilisant uniquement quelques **fonctionnalités ou colonnes de l’ensemble d’entraînement**. 

+ Après avoir ajusté le nouvel arbre et lié les erreurs d’entraînement et de test - aïe - tout d’un coup, vous avez de grosses erreurs dans les deux cas. 

+ L’arbre simple n’est pas capable de capturer très bien la complexité de l’ensemble de formation ou de test. 

+ Vous sous-ajustez les données. Nous appelons cet effet « biais élevé ».

####**6. Le compromis biais-variance**

+ Ce graphique montre la relation entre ces deux observations. 

+ Pour les modèles très simples, nous observons généralement un biais élevé et une faible variance. 

+ Pour les modèles trop complexes, nous observons un biais très faible, mais une variance élevée. 

+ Ce compromis dans la complexité est appelé le compromis biais-variance. 

+ Vous devez trouver le bon équilibre entre la complexité du modèle sans surajuster ou sous-ajuster les données : le point idéal au centre.

####**7. Détection du surajustement**

+ Qu’est-ce que tout cela signifie en pratique? 

+ Comme vous l’avez appris dans la leçon précédente, ***la validation croisée est idéale pour utiliser de nombreux ensembles de formation et de test pendant le développement du modèle sans même toucher à l’ensemble de test final***.

+ Comme l’erreur de CV est beaucoup plus élevée que l’erreur d’apprentissage, nous pouvons en déduire que votre modèle souffre de surajustement. 

+ Vous devez réduire la complexité de l’arborescence, par exemple, en diminuant le nombre d’entités ou la profondeur de l’arborescence.

####**8. Détection du sous-équipement**

+ Supposons maintenant que vous l’ayez fait et que vous réadaptiez votre modèle à l’ensemble de l’ensemble d’entraînement. 

+ Encore une fois, vous calculez votre erreur absolue moyenne dans l’échantillon et vous constatez qu’elle est très élevée. 

+ C’est un indicateur clair que votre modèle souffre maintenant d’un sous-équipement. Vous avez fait trop de simplification et vous devriez augmenter la complexité du modèle.

###**Réglage des hyperparamètres**

####**1. Réglage des hyperparamètres**
+ Bienvenue!

####**2. Hyperparamètres**

+ Comme nous en avons discuté plus tôt dans le cours, il existe plusieurs boutons que vous pouvez tourner pour **influencer la forme et la complexité de vos arbres**. 

+ Certains boutons donnent des arbres plus grands ou plus complexes, tandis que d’autres forcent les arbres à rester petits et peu profonds. 

+ **Ces boutons sont appelés « hyperparamètres »**. 

+ Ils contrôlent la complexité d’un modèle et doivent être spécifiés avant la formation du modèle. 

+ Pour les arbres de décision, nous avons les trois paramètres suivants : 

  + **min_n**, qui est le nombre minimum d’échantillons requis pour diviser un nœud,

  + **tree_depth**, qui représente la profondeur maximale d’un arbre, 

  + et **cost_complexity**, qui est un nombre de pénalités qui punit la taille de l’arbre.

####**3. Pourquoi le réglage ?**

+ Lorsque vous ne spécifiez aucun de ces hyperparamètres, parsnip choisit les valeurs par défaut suivantes : 
  + échantillons minimum par nœud de 20, profondeur maximale de 30 et cost_complexity de 0-point-01. 
  
  + Ceux-ci sont choisis pour qu’ils fonctionnent bien dans de nombreux cas. 

  + Cependant, ils peuvent ne pas être le meilleur choix pour tous les jeux de données.

  + L’objectif du réglage est maintenant de trouver l’ensemble des hyperparamètres qui donnent à votre modèle les meilleures performances possibles.

####**4. Réglage avec des tidymodels à l’aide du package de réglage**

+ **Tidymodels** dispose d’une boîte à outils contenant des fonctions qui facilitent grandement le réglage des hyperparamètres : **Le package tune**. 

+ La première étape consiste à créer un ensemble de valeurs d’hyperparamètres, appelé grille, c’est-à-dire le pool dans lequel nous recherchons le meilleur candidat.

+ La deuxième étape consiste à créer une spécification de modèle factice dans laquelle nous étiquetons les hyperparamètres pour le réglage.

####**7. Réglage avec des modèles bien rangés**

####**6. Courbe ROC (Receiver-operating-characteristic)**

+ ***La courbe ROC, acronyme de receiver-operating-characteristic, visualise la performance d’un modèle de classification sur tous les seuils de probabilité possibles***. 

+ Pour chaque seuil unique, un point qui représente le taux ou la sensibilité positive réel et le taux de faux positifs ou un moins la spécificité est ajouté au graphique.

####**7. Courbe ROC et AUC**


+ Une courbe ROC idéale tomberait dans le coin supérieur gauche du graphique, avec un taux ou une sensibilité positive réelle de 1 et un taux de faux positifs ou 1 moins la spécificité de 0. 

+ La performance globale d’un classificateur, résumée sur tous les seuils possibles, est donnée par l’aire sous cette courbe (AUC).

####**8. Aire sous la courbe ROC**

+ Une AUC du point zéro cinq correspond à une courbe ROC le long de la diagonale et indique qu’un modèle ne fonctionne pas mieux que le hasard. 

+ Une AUC de un correspond à un modèle qui classe parfaitement chaque exemple, tandis qu’une AUC de zéro signifie que votre modèle classe mal chaque observation. 

+ Ainsi, les courbes ROC et l’AUC sont utiles pour comparer différents classificateurs, car elles prennent en compte tous les seuils possibles.

####**1. Arbres ensachés**

+ Les arbres de décision sont faciles à comprendre et à interpréter, mais souvent un petit changement dans les données peut entraîner une série très différente de divisions et un modèle très différent. 

+ C’est l’un des principaux inconvénients des arbres de décision : la forte variance que vous avez observée dans les exercices.

####**2. Beaucoup de têtes valent mieux qu’une**

+ Une solution à cela est la sagesse de la foule: 
  + Les connaissances collectives de nombreuses personnes dépassent généralement les connaissances d’un seul individu. 
  
  + Cela soulève l’idée de ce que l’on appelle des « modèles d’ensemble », ce qui signifie que vous utilisez plusieurs modèles et leurs prédictions ensemble au lieu d’un seul modèle.

####**3. Bootstrap et agrégation**

+ **Que signifie le terme « arbres ensachés »?** 
+ ***L’ensachage est une méthode d’ensemble et est un raccourci pour Bootstrap Aggregation****. 

+ L’amorçage signifie simplement échantillonner des lignes au hasard à partir du jeu de données d’apprentissage, avec remplacement. 

+ Lorsque vous tirez des échantillons avec remplacement, cela signifie que vous dessinerez un seul exemple d’entraînement plus d’une fois. 

+ Il en résulte une version modifiée de l’ensemble d’apprentissage où certaines lignes sont représentées plusieurs fois et d’autres sont absentes. 

+ Cela vous permet de générer de nouvelles données similaires aux données avec lesquelles vous avez commencé. 

+ En faisant cela, vous pouvez adapter de nombreux modèles différents, mais similaires. 

+ L’agrégation est effectuée en utilisant la prédiction moyenne de tous les modèles comme prédiction de régression finale, ou le vote majoritaire en classification.

####***4. Étape 1 : Bootstrap et train**

+ L’ensachage fonctionne de la manière suivante : 
  + Tout d’abord, vous tirez un échantillon avec remplacement à partir de l’ensemble d’entraînement d’origine 
  + Ensuite, vous entraînez un modèle d’arbre de décision à l’aide de cet ensemble d’apprentissage échantillonné. 
  + Répétez ces étapes autant de fois que vous le souhaitez 
    - cela pourrait être 10 fois, 100 fois ou 1000. 
    
+ En règle générale, plus il y a d’arbres, meilleur est le modèle, mais plus vous avez besoin de temps d’entraînement. Cet exemple montre un ensemble de trois arbres ensachés.

####**5. Étape 2 : Agréger**

+ Maintenant, disons que vous avez trois arbres bootstrapés qui composent votre ensemble. 

+ Pour générer une prédiction à l’aide d’un modèle d’arbre en sac, vous générez des prédictions à partir de chacun des arbres, puis vous agrégez simplement les prédictions pour obtenir une prédiction finale. 

+ La prédiction en sac, ou « ensemble », est la prédiction moyenne ou le vote majoritaire à travers les arbres bottés. 

+ L’ensachage peut réduire considérablement la variance des modèles instables tels que les arbres, ce qui améliore les performances.

###**Forêt aléatoire**

####**1. Forêt aléatoire**
+ Bienvenue! Maintenant, vous savez que **les arbres ensachés sont une grande amélioration par rapport aux arbres à décision unique**. 

+ Vous avez peut-être entendu parler de **forêts aléatoires, qui sont une amélioration par rapport aux arbres ensachés**.

####**2. Forêt aléatoire**
+ **Les forêts aléatoires sont particulièrement adaptées aux données de grande dimension**. 

+ En raison de leur facilité d’utilisation et de leurs performances prêtes à l’emploi, la forêt aléatoire est un algorithme d’apprentissage automatique très populaire et est implémenté dans une variété de packages, comme **le package ranger ou randomForest**. 

+ **Tidymodels** vous couvre : 
  + **la fonction rand_forest() du paquet Parsnip** fournit une interface pour ces implémentations.

####**3. Idée**

+ L’idée de base derrière **la forêt aléatoire est identique à l’ensachage**

  - ***les deux sont des ensembles d’arbres formés sur des échantillons bootstrapés des données d’apprentissage***. 
  
+ Cependant, dans l’algorithme de forêt aléatoire, il y a un léger ajustement à la façon dont les arbres de décision sont construits qui conduit à de meilleures performances. 

+ ***La principale différence est que lors de l’entraînement des arbres qui composent l’ensemble, nous ajoutons un peu de hasard supplémentaire au modèle - d’où le nom, forêt aléatoire***. 

+ À chaque scission de l’arborescence, plutôt que de considérer toutes les caractéristiques, ou variables d’entrée, pour la division, nous échantillonnons un sous-ensemble de ces caractéristiques ou prédicteurs et considérons uniquement ces quelques variables comme candidates à la division.

####**4. L’intuition**
+ Aiguisons notre intuition en utilisant cette image. 

+ Dans ce cas, il y a quatre arbres, et dans chaque division, différentes variables ou caractéristiques prédictives sont utilisées pour effectuer la division suivante. 

+ Chaque arbre donne ensuite un vote pour une classe, et le vote majoritaire est la prédiction finale de classe. 

+ Comment l’utilisation de moins de prédicteurs peut-elle conduire à de meilleures performances ? 
+ Eh bien, ajouter ce peu plus de hasard conduit à une collection d’arbres qui sont encore plus décorrélés (ou plus différents) les uns des autres. 

+ Ainsi, la forêt aléatoire améliore l’ensachage en réduisant la corrélation entre les arbres échantillonnés.

###**Renforcement du gradient**

####**1. Renforcement du gradient**
+ Bienvenue! Récapitulons et renforçons ce que nous savons sur la stimulation jusqu’à présent.

####**2. Récapitulatif : booster**

+ Dans le boosting, les apprenants faibles comme les arbres de décision avec une seule division sont utilisés, qui ne fonctionnent que légèrement mieux qu’un hasard aléatoire. 

+ Boosting se concentre sur l’addition séquentielle de ces apprenants faibles et le filtrage des observations qu’un apprenant obtient correcte à chaque étape. 

+ Fondamentalement, l’accent est mis sur le développement de nouveaux apprenants faibles pour gérer les observations difficiles restantes à chaque étape. 

+ L’un des tout premiers algorithmes de boost développés a été Adaboost. Le renforcement du gradient a amélioré certaines des fonctionnalités d’Adaboost pour créer un algorithme plus fort et plus efficace.

####**3. Comparaison**
+ Adaboost utilise les souches de décision comme apprenants faibles. 

+ Les souches de décision ne sont rien d’autre que des arbres de décision avec une seule division. 

+ Il attache également du poids aux observations, ajoutant plus de poids aux observations « difficiles à classer » et moins de poids à celles qui sont faciles à classer. 

+ L’amplification de gradient utilise des arbres de décision courts et moins complexes au lieu de souches de décision. 

+ Au lieu de peser les observations, il utilise une fonction d’erreur appelée fonction de perte pour mesurer à quelle distance elle se trouve. 

+ Étant donné que l’optimisation de la fonction de perte est effectuée à l’aide de la descente de gradient, la méthode est appelée « amplification de gradient ».

####**4. Avantages et inconvénients de la stimulation**

+ L’une des raisons pour lesquelles le boost est si populaire est que s’il est réglé correctement, les performances sont souvent meilleures que tout autre algorithme de votre boîte à outils. 

+ Un modèle boosté optimisé peut surpasser même l’état de l’art en deep learning sur de nombreux ensembles de données. 

+ La mise en avant est une bonne option pour les ensembles de données déséquilibrés. 

+ Dans des applications telles que la falsification ou la détection de fraude, les classes seront presque certainement déséquilibrées, où le nombre de transactions authentiques sera énorme par rapport aux transactions non authentiques. 

+ Un problème que nous pouvons rencontrer dans les arbres de décision de gradient stimulant mais pas les forêts aléatoires est le surajustement dû à l’ajout de trop d’arbres. 

+ Dans les forêts aléatoires, l’ajout de trop d’arbres ne provoquera pas de surajustement. 

+ La précision du modèle ne s’améliore pas après un certain point, mais aucun problème de surajustement n’est rencontré. 

+ Selon la façon dont vous ajustez l’hyperparamètre du taux d’apprentissage, l’apprentissage dans les ensembles boostés peut être lent, d’autant plus qu’il s’agit d’un algorithme itératif et non parallèle.
