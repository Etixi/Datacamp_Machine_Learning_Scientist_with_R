####**1. Régression linéaire bayésienne**


+ Maintenant que nous avons passé en revue **les caractéristiques d’une régression fréquentiste**, examinons **comment estimer cette même régression à l’aide de méthodes bayésiennes.**

####**2. Pourquoi utiliser des méthodes bayésiennes ?**

+ Comme nous l’avons appris dans la dernière leçon, si nous voulons faire des inférences sur les valeurs réelles des paramètres, les valeurs $p$ et la régression fréquentiste nous échouent. 

+ **L’estimation bayésienne est une solution à ce problème**. 

+ ***Avec les méthodes bayésiennes, la probabilité utilisée pour la régression fréquentiste, et ce que l’on appelle un a priori, forment une distribution postérieure***. 

+ Les détails de ce processus dépassent la portée de ce cours. 

+ ***Le point clé est que les méthodes bayésiennes échantillonnent à partir de cette distribution postérieure, et nous pouvons ensuite créer des résumés des distributions pour faire des inférences de paramètres.***

+ L’utilisation des résumés nous permet de faire des inférences sur les valeurs que les paramètres peuvent prendre.

####**3. Le paquet rstanarm**
+ Pour estimer les modèles de régression bayésiens dans ce cours, nous utiliserons le package rstanarm. 

+ ***rstanarm est une interface avec Stan, qui est un langage de programmation pour l’inférence bayésienne.*** 

+ ***Le package rstanarm offre une interface de haut niveau avec des scripts Stan pré-écrits pour les modèles courants, comme la régression linéaire.***

+ C’est parce que les échantillons d’estimation bayésienne de la distribution postérieure. 

+ Cela signifie qu’au lieu d’une estimation ponctuelle et d’une statistique de test, nous obtenons une distribution de valeurs plausibles pour les paramètres, et la section des estimations résume ces distributions.

####**7. rstanarm résumé: Diagnostics**


+ Dans la section diagnostic, la statistique la plus importante à laquelle il faut prêter attention est le R-hat. 

+ Contrairement à la régression fréquentiste où il y a toujours une solution utilisant les moindres carrés ordinaires, dans les modèles bayésiens, nous devons vérifier pour nous assurer que le modèle a convergé. 

+ Si un modèle est convergé, les estimations des paramètres sont stables. Sinon, nos résultats ne seront pas fiables. 

+ Dans l’estimation bayésienne, les distributions postérieures sont échantillonnées en groupes, appelés chaînes. 

+ En comparant la variance à l’intérieur des chaînes à la variance entre les chaînes, nous pouvons mesurer la stabilité de nos estimations. C’est la statistique R-hat. 

+ En général, nous voulons que toutes les valeurs R-hat soient inférieures à 1,1 afin de conclure que le modèle a convergé, comme dans cet exemple.

###**Comparaison des approches bayésiennes et fréquentistes**

####**1. Comparaison des approches bayésiennes et fréquentistes**
+ Vous avez peut-être remarqué dans la dernière leçon que nos estimations utilisant **la fonction stan_glm** étaient très similaires à nos estimations de paramètres lors de l’utilisation de **la fonction lm**. 

+ Si c’est le cas, vous vous demandez probablement pourquoi nous devons même nous embêter avec Bayesian. 

+ Nous en avons parlé un peu, mais cela vaut la peine de prendre un peu de temps maintenant pour vraiment comprendre les différences entre les camps fréquentistes et bayésiens.

####**2. Les mêmes paramètres!**


+ Lorsque nous examinons les résultats des **modèles lm et stan_glm** que nous avons estimés précédemment, **nous constatons que les estimations et les erreurs-types sont très similaires**. 

+ Cependant, la sortie du modèle bayésien **ne contient pas de statistiques de test ou de valeurs p**.

####**3. Fréquentiste vs bayésien**


+ ***Cela représente la différence fondamentale entre les fréquentistes et les bayésiens.*** 

+ En bref, ***les fréquentistes supposent que les paramètres du modèle sont fixes et que les données sont aléatoires, tandis que les bayésiens supposent que les données sont fixes et que les paramètres du modèle sont aléatoires.***

+  Cela peut être vu dans l’interprétation d’une valeur de p. 

+ La valeur de p est la probabilité d’observer des données qui donnent lieu à une statistique de test aussi grande, si l’hypothèse nulle est vraie. 

+ En d’autres termes, ***étant donné un ensemble de valeurs de paramètres vraies (l’hypothèse nulle), quelle est la probabilité d’observer un jeu de données aléatoire qui aboutit à une statistique de test aussi grande.***

+ En revanche, ***les bayésiens supposent que les données collectées sont fixes, et qu’il existe à la place une distribution des valeurs de paramètres possibles qui donnent lieu aux données.*** 

+ ***En d’autres termes, les bayésiens s’intéressent à la détermination de la plage de valeurs des paramètres qui donneraient lieu à leur ensemble de données observées.***

####**4. Évaluation des paramètres bayésiens**

+ Cependant, il est souvent souhaitable de disposer d’une méthode permettant d’évaluer si un paramètre estimé à l’aide de méthodes bayésiennes est significatif ou significatif. 

+ Pour cela, **nous utiliserons des intervalles crédibles**. 

+ **Un intervalle crédible semblera très similaire à un intervalle de confiance**.

+ **Un intervalle de confiance nous indique la probabilité qu’une plage contienne la valeur vraie**. 

+ Cependant, **l’intervalle de confiance ne peut rien nous dire quelle est la probabilité de valeurs spécifiques pour le paramètre d’intérêt**. 

+ Cela peut sembler couper les cheveux en quatre, mais c’est une distinction importante à faire pour s’assurer que nous faisons les bonnes inférences. 

+ En revanche, **des intervalles crédibles nous indiquent à quel point les valeurs sont probables**. 

+ Cela nous permet de faire des inférences sur les valeurs réelles des paramètres, plutôt que sur les limites des plages.

####**5. Créer des intervalles crédibles**

+ **Dans rstanarm**, nous pouvons facilement **calculer des intervalles crédibles en utilisant la fonction 'posterior_interval'**. 

+ Par défaut, rstanarm fournit l’intervalle de crédibilité à 90%, mais nous pouvons créer des intervalles crédibles à 95 ou 50% en fournissant l’intervalle souhaité à l’argument prob.

####**6. Confiance vs intervalles crédibles**

+ Ces intervalles ressemblent beaucoup à l’intervalle de confiance correspondant. Voici l’intervalle de confiance à 95 % pour le paramètre mom_iq dans notre modèle fréquentiste. 

+ Cela nous indique qu’il y a 95% de chances que la plage de 0,49 à 0,72 contienne la valeur réelle. 

+ Mais nous nous intéressons à la probabilité que la valeur tombe entre deux points, pas à la probabilité que les deux points capturent la vraie valeur. 

+ C’est ce que nous donne l’intervalle crédible. 

+ Comme vous pouvez le voir, ces intervalles donnent des plages très similaires.

+ Cela peut souvent être le cas, mais les inférences sont très différentes. 

+ Dans le scénario bayésien, nous pouvons demander quelle est la probabilité que le paramètre soit compris entre 0,60 et 0,65, et nous voyons une probabilité de 31% que la valeur réelle soit dans cette plage. 

+ Comment ferions-nous quelque chose de similaire avec des intervalles de confiance d’un modèle fréquentiste? On peut pas. 

+ Seules les méthodes bayésiennes nous permettent de faire des inférences sur les valeurs réelles du paramètre.

###**Qu’y a-t-il dans un modèle bayésien?**

####**1. Qu’y a-t-il dans un modèle bayésien?**

+ Dans le chapitre précédent, nous avons appris à estimer une régression linéaire bayésienne. 

+ Dans ce chapitre, nous allons apprendre **comment ces modèles peuvent être modifiés**. 

+ ***Contrairement à une régression fréquentiste, il existe plusieurs façons de modifier le processus d’estimation***. 

+ Ceci est important car ces facteurs auront finalement un impact sur vos résultats. 

+ Nous allons commencer par examiner l’échantillonnage du postérieur qui crée les résumés de distribution que nous voyons dans la sortie.

####**2. Distributions postérieures**


+ Comme nous en avons brièvement parlé au chapitre précédent, **la distribution postérieure est échantillonnée en groupes, appelés chaînes**. 

+ **Chaque échantillon de la chaîne est appelé itération**. 

+ **Les chaînes commencent à des endroits aléatoires**. 

+ Au fur et à mesure de l’échantillonnage, **la chaîne se déplace vers la zone où la combinaison de la probabilité et du préalable indique une forte probabilité que la valeur réelle du paramètre réside.** 

+ ***Plus il y a d’itérations dans une chaîne, plus les échantillons de la distribution postérieure seront grands.*** 

+ ***Cela signifie que les résumés sont directement influencés par la longueur de la chaîne, car des échantillons plus grands permettent des estimations plus robustes de ces statistiques sommaires.***

####**3. Échantillonnage de la distribution postérieure**


+ Voici un exemple de ce processus. **C’est ce qu’on appelle un tracé de trace**. 

+ ***Il affiche la valeur du paramètre à chaque itération pour chaque chaîne.*** 

+ Nous pouvons voir que chaque chaîne commence à un endroit différent, mais elles convergent toutes vers la même zone. 

+ C’est la convergence dont nous avons parlé avec le paramètre R-hat. 

+ La convergence est importante car elle garantit des estimations stables. 

+ Nous pouvons voir que ce modèle n’a pas convergé au début, car les chaînes sont à différents endroits et non horizontales, ce qui signifie que les estimations ne sont pas stables.

####**4. Échantillonnage de la distribution postérieure**


+ Étant donné que le modèle n’a pas convergé au début, nous écartons ces itérations, ne laissant que les itérations convergées pour constituer notre distribution postérieure finale. 

+ Ici, nous pouvons voir qu’en utilisant uniquement les 1 000 itérations finales, toutes les chaînes sont entièrement mélangées. 

+ Les itérations qui sont rejetées sont connues sous le nom d’échauffement ou de burn-in. 

+ Par défaut, le paquet rstanarm estime 4 chaînes.

####**2. Transitions divergentes**

+ Le premier message d’erreur dont nous parlerons concerne les transitions divergentes. 

+ Cela se produit lorsque la taille entre les étapes de l’estimateur est trop grande. 

+ **La documentation de Stan décrit cela comme une descente d’une colline escarpée.** 

+ Faire un pas trop grand peut vous faire tomber. 

+ Mais vous pouvez faire de plus petits pas et le faire en toute sécurité, même si cela prend plus de temps. 

+ Nous pouvons contrôler la taille des pas dans Stan en envoyant une liste à l’argument 'control' qui spécifie un adapt_delta. 

+ L’argument adapt_delta peut être compris entre 0 et 1 et est défini sur 0-point-9-5 par défaut dans rstanarm. 

+ En augmentant le adapt_delta, nous pouvons réduire la taille des pas et, dans de nombreux cas, résoudre l’erreur de transitions divergentes.

+ Le deuxième message d’erreur que vous pouvez parfois rencontrer est celui indiquant que la profondeur maximale de l’arbre a été atteinte. 

+ Dans l’algorithme d’échantillonnage utilisé par Stan et donc rstanarm, l’échantillonneur cherche un endroit pour « U-Turn » dans une série de branches possibles. 

+ Encore une fois, les détails de ce processus dépassent la portée de ce cours. 

+ Ce qu’il est important de savoir ici, c’est que si l’échantillonneur atteint la profondeur maximale de l’arbre avant de trouver un bon endroit pour « U-Turn », alors l’échantillon termine l’itération avant de trouver un bon point d’arrêt. 

+ Par conséquent, la distribution postérieure n’est pas échantillonnée aussi efficacement qu’elle devrait l’être. 

+ Comme pour adapt_delta, nous pouvons contrôler la profondeur maximale de l’arbre dans Stan en envoyant une liste à l’argument de contrôle qui spécifie un max_treedepth. 

+ Le max_treedepth est défini sur 10 par défaut dans rstanarm. En augmentant le max_treedepth, nous permettons à la recherche d’un bon endroit pour « U-Turn ».

####**4. Réglage de l’estimation**

###**Utilisation de la statistique R au carré**

####**1. Utilisation de la statistique R au carré**

+ Jusqu’à présent dans le cours, nous avons parlé de la façon d’estimer et de modifier un modèle. 

+ Dans les deux derniers chapitres, nous examinerons comment les modèles que nous estimons peuvent être évalués et utilisés pour faire des prédictions. 

+ Dans ce chapitre, nous nous concentrerons sur la première partie : l’évaluation des modèles. 

+ C’est une partie cruciale de la modélisation de la construction, car si notre modèle ne correspond pas, nous ne serons pas en mesure de faire de très bonnes prédictions. 

+ Nous commencerons par utiliser l’une des mesures les plus courantes de l’ajustement du modèle dans la régression linéaire : **la statistique r-carrée.**

####**2. Qu’est-ce que R au carré?**

+ La statistique R au carré est une mesure de la capacité des variables indépendantes du modèle à prédire la variable dépendante. 

+ Plus précisément, la statistique R au carré mesure la proportion de variance dans la variable dépendante qui peut être expliquée par les variables indépendantes. 

+ Comme pour toutes les proportions, le carré R varie de 0 à 1, 0 représentant aucune variance expliquée et 1 représentant toute la variance expliquée, ou un modèle déterministe. 

+ Pour cette raison, le R au carré est également connu sous le nom de coefficient de détermination. 

+ Le R au carré est calculé comme suit : 1 moins la somme des carrés résiduels (appelée somme résiduelle des carrés) divisée par la somme des écarts au carré des données par rapport à la moyenne (appelée somme totale des carrés).

####**3. Qu’est-ce que R au carré?**
+ En d’autres termes, nous prenons la valeur observée pour une observation et soustrayons la valeur prédite, la mettons au carré, puis additionnons cela sur toutes les observations. C’est le numérateur.

####**4. Qu’est-ce que R au carré?**

+ Dans le dénominateur, nous prenons la valeur observée pour une observation, soustrayons la moyenne de la valeur observée, quadratissons cela et additionnons toutes les observations. 

+ Cette variance du rapport des résidus par rapport à la variance totale est ce qui détermine le R au carré.

####**5. Calcul de la statistique R au carré**

+ Dans une régression fréquentiste, nous pouvons obtenir le R au carré en estimant un modèle avec la fonction lm() et en enregistrant un résumé de l’objet. 

+ Cet objet contient un 'r.squared', que nous pouvons extraire pour voir. 

+ Cependant, nous pouvons également calculer cela à la main. 

+ Nous pouvons définir la somme résiduelle des carrés comme la variance des résidus de lm_model et la somme totale des carrés comme la somme de la variance des résidus de lm_model et de la variance des valeurs prédites de lm_model. 

+ En prenant 1 moins la somme résiduelle du carré divisée par la somme totale des carrés, nous obtenons la même valeur qui a été enregistrée dans le lm_summary.

####**6. La statistique R carré d’un modèle bayésien**

+ Dans rstanarm, le R carré n’est pas enregistré dans l’objet summary comme dans la fonction summary lm(). 

+ Cependant, nous pouvons toujours calculer le R au carré à la main en utilisant exactement les mêmes formules que pour la régression fréquentiste. 

+ Nous définissons la somme résiduelle des carrés et la somme totale des carrés, puis nous les utilisons pour calculer la valeur R au carré. 

+ En comparant cette valeur à la valeur que nous avons obtenue dans la régression fréquentiste, nous voyons qu’elles sont presque identiques.

####**1. Vérifications a posteriori du modèle prédictif**

+ La plus importante d’entre elles est une classe d’analyses appelées vérifications de modèles prédictifs postérieurs, qui ne peuvent être calculées que lorsqu’un processus d’estimation bayésien est utilisé.

####**2. Utilisation des distributions postérieures**

+ Comme son nom l’indique, les vérifications de modèles prédictifs postérieurs utilisent les distributions postérieures. 

+ Rappelez-vous notre modèle prédisant le score de QI d’un enfant à partir du score de QI de sa mère.

+ Comme nous pouvons le voir, les valeurs changent à chaque itération, ce qui est attendu car nous échantillonnons aléatoirement à partir de la distribution postérieure.

####**3. Prédictions postérieures**

+ Nous pouvons utiliser ces tirages postérieurs pour calculer le score prédit pour nos données. 

+ Par exemple, nous pouvons calculer les scores prédits à l’aide des valeurs des paramètres à l’itération 1, un autre ensemble de scores prédits à l’aide des valeurs des paramètres à l’itération 2, etc. 

+ En fin de compte, comme nous avons 4 000 itérations, nous pouvons calculer 4 000 scores prédits pour chaque observation. 

+ Nous pouvons obtenir toutes les valeurs prédites pour toutes les observations en utilisant la fonction posterior_linpred. 

+ Cela renvoie une matrice comportant une ligne pour chaque itération et une colonne pour chaque observation.

+ Étant donné que ces scores sont générés à l’aide des paramètres du modèle, ces scores prédits forment une distribution de ce à quoi nos scores observés devraient ressembler - si le modèle spécifié était correct. 

+ Ainsi, les écarts par rapport à ces distributions prédictives sont une indication d’un mauvais ajustement.

####**4. Comparaison des distributions de scores**

+ Une façon de rechercher des écarts consiste à comparer la distribution des scores prédits dans une itération à la distribution des scores observés. 

+ Par exemple, nous pouvons obtenir les scores prédits pour les première et deuxième itérations en tirant les première et deuxième lignes du retour de la matrice par posterior_linpred.

####**2. Le paquet de toilettes**
+ Lorsque nous estimons un modèle avec rstanarm, nous pouvons utiliser le package loo pour les comparaisons de modèles. 

+ LOO signifie « leave-one-out » qui est une variété d’une classe d’outils de comparaison de modèles plus largement connus sous le nom de validation croisée. 

+ Le package loo n’utilise pas de véritable validation croisée. 

+ Au lieu de cela, l’algorithme se rapproche de la validation croisée de non-sortie. 

+ Les détails de la façon exacte dont cela est fait dépassent la portée de ce cours, mais si vous voulez en savoir plus, la documentation du package Loo fournit de nombreuses ressources pour l’exploration. 

+ Dans cette leçon, nous allons plutôt nous concentrer sur la façon dont le package de toilettes peut être utilisé pour comparer des modèles et interpréter la sortie.

####**3. Utilisation des toilettes sur un seul modèle**

+ Pour afficher les estimations LOO d’un modèle, nous envoyons simplement un modèle estimé comme argument à **la fonction loo()**. 

+ Cela nous donne plusieurs informations. Tout d’abord, nous voyons que les estimations ont été calculées à partir d’une matrice log-vraisemblance qui comporte 4000 lignes (le nombre total d’itérations dans notre postérieur) et 434 colonnes (le nombre total d’observations dans l’ensemble de données kidiq). 

+ Nous obtenons l’estimation LOO, appelée elpd_loo, le nombre effectif de paramètres dans le modèle, p_loo, et l’estimation LOO convertie en une échelle de déviance, looic. 

+ L’échelle de déviance est simplement moins deux fois l’estimation de la LOO, mais elle est plus courante dans certains domaines. 

+ Enfin, nous recevons quelques diagnostics de l’algorithme d’approximation. 

+ Cependant, ces chiffres ne sont pas très utiles isolément. Par exemple, que signifie une valeur epld_loo de moins 1878 points 5? Cette valeur n’a vraiment de sens que par rapport aux valeurs des modèles concurrents.

####**4. Comparaisons de modèles avec loo**

+ Disons que nous avons deux modèles possibles. Dans le premier modèle, nous prédisons les scores de QI d’un enfant à partir du QI de sa mère. 

+ Dans le deuxième modèle, nous prédisons le score de QI de l’enfant non seulement à partir du QI de sa mère, mais aussi du fait que sa mère ait obtenu ou non son diplôme d’études secondaires, et de l’interaction entre le QI de la mère et l’achèvement des études secondaires de la mère. 

+ Nous voulons savoir quel modèle fait un meilleur travail de prédiction du score de QI de l’enfant. 

+ Pour ce faire, nous pouvons enregistrer les estimations de loo à partir des modèles 1 prédicteur et 2 modèles de prédiction, puis utiliser la fonction « comparer ».

####**5. Comparaisons de modèles avec les toilettes**

+ La fonction de comparaison fournit l’utilisation de la différence dans les estimations de loo, ainsi qu’une erreur type de la différence. 

+ Un score de différence positif comme dans cet exemple, signifie que le deuxième modèle est favorisé (le modèle avec les deux prédicteurs), alors qu’un score négatif indiquerait une préférence pour le premier modèle. L’erreur-type nous aide à décider si la différence est significative. 

+ En règle générale, si la valeur absolue de la différence est inférieure à l’erreur-type, les modèles ne fonctionnent pas différemment, et nous devrions choisir le modèle le plus simple (celui avec moins de paramètres) car il est plus parcimonieux. 

+ Si la valeur absolue de la différence est supérieure à l’erreur type, nous préférerons le modèle indiqué par le test. 

+ Dans cet exemple, parce que la différence de 6,1 est positive et supérieure à l’erreur-type de 3,9, nous choisirions le deuxième modèle avec les deux prédicteurs.
