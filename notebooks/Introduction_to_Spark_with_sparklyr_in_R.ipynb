+ Une gnralisation de l'ide prcdente consiste  avoir plusieurs seuils ; c'est--dire que vous divisez une variable continue en "buckets" (ou "bins"), tout comme le fait un histogramme.

####**Comparaison des performances des modles**

+ Le trac vous donne une bonne ide des endroits o le modle est performant et de ceux o il ne l'est pas. 

+ Parfois, il est intressant d'avoir une statistique qui vous donne un score pour le modle. 

+ Ainsi, vous pouvez quantifier la qualit d'un modle et effectuer des comparaisons entre plusieurs modles. 

+ Une statistique courante est l'erreur quadratique moyenne (parfois abrge en "RMSE"), qui lve simplement les rsidus au carr, puis prend la moyenne, puis la racine carre. 

+ Un petit score RMSE pour un ensemble de donnes donn implique une meilleure prdiction. 

+ (Par dfaut, vous ne pouvez pas comparer diffrents ensembles de donnes, mais uniquement diffrents modles sur le mme ensemble de donnes. Il est parfois possible de normaliser les jeux de donnes afin de pouvoir les comparer).

*Ici, vous allez comparer les arbres boosts par le gradient et les modles de fort alatoire.*

####**Instructions**

*both_responses, contenant l'anne prdite et l'anne relle de la piste  partir des deux modles, a t prdfini comme un tibble local.*

+ Crez un ensemble de donnes de la somme des carrs des rsidus.

+ Ajoutez une colonne rsiduelle, gale  la rponse prdite moins la rponse relle.

+ Regroupez les donnes par modle.

+ Calculez une statistique sommaire, rmse, gale  la racine carre de la moyenne des rsidus au carr.

```r
# both_responses has been pre-defined
both_responses

both_responses %>%
  # Add a residual column
  mutate(
    residual = predicted - actual
  ) %>%
  # Group by model
  group_by(model) %>%
  # Calculate the root mean square error
  summarize(
    rmse = sqrt(mean(residual ^ 2))
  )
```
