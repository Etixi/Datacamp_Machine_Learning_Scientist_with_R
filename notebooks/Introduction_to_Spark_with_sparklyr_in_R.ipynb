####**Big data, tiny tibble**

+ Dans le dernier exercice, lorsque vous avez copié les données vers Spark, **copy_to() a renvoyé une valeur**. 

+ Cette valeur de retour est un type spécial de **tibble()** qui ne contient pas de données propres. 

+ Pour expliquer cela, vous devez connaître un peu la façon dont les paquets tidyverse stockent les données. 

+ Les tibbles sont généralement juste une variante des data.frames qui ont une méthode d'impression plus agréable. 

+ Cependant, dplyr leur permet également de stocker des données provenant d'une source de données distante, comme des bases de données, et - comme c'est le cas ici - Spark. 

+ Pour les ensembles de données distants, l'objet tibble stocke simplement une connexion aux données distantes. 

+ Nous y reviendrons plus en détail plus tard, mais le point important pour l'instant est que même si vous avez un grand ensemble de données, la taille de l'objet tibble est petite.

+ Du côté de Spark, les données sont stockées dans une variable appelée DataFrame. 

+ Il s'agit d'un équivalent plus ou moins direct du type de variable data.frame de R. 

+ (Bien que les types de variables de colonnes soient nommés légèrement différemment - par exemple, les colonnes numériques sont appelées colonnes DoubleType). 

+ Tout au long du cours, le terme data frame sera utilisé, sauf si une clarification est nécessaire entre data.frame et DataFrame. 

+ Puisque ces types sont également analogues aux tables des bases de données, le terme table sera parfois utilisé pour décrire ce type de données rectangulaires.

+ L'appel de tbl() avec une connexion Spark, et une chaîne de caractères nommant le cadre de données Spark retournera le même objet tibble que celui retourné lorsque vous avez utilisé **copy_to()**.

+ Un outil utile que vous verrez dans cet exercice est la fonction object_size() du paquet pryr. Elle vous montre la quantité de mémoire que prend un objet.

####**Explorer la structure des tibbles**

+ Si vous essayez d'imprimer un tibble qui décrit des données stockées dans Spark, une certaine magie doit se produire, puisque le tibble ne conserve pas de copie des données elles-mêmes. 

+ La magie réside dans le fait que la méthode d'impression utilise votre connexion Spark, copie une partie du contenu vers R, et affiche ces valeurs comme si les données avaient été stockées localement. 

+ Comme vous l'avez vu plus tôt dans le chapitre, la copie des données est une opération lente, donc par défaut, seules 10 lignes et autant de colonnes qui tiennent sur l'écran, sont imprimées.

+ Vous pouvez modifier le nombre de lignes qui sont imprimées en utilisant l'argument n de **print()**. 

+ Vous pouvez également modifier la largeur du contenu à afficher en utilisant l'argument width, qui est spécifié comme le nombre de caractères (et non le nombre de colonnes). 

+ Une astuce intéressante consiste à utiliser **width = Inf** pour imprimer toutes les colonnes.

+ **La fonction str()** est généralement utilisée pour afficher la structure d'une variable. 

+ Pour les data.frames, elle donne un bon résumé avec le type et les premières valeurs de chaque colonne. 

+ Cependant, pour les tibbles qui ont une source de données distante, **str()** ne sait pas comment récupérer les données. 

+ Cela signifie que si vous appelez str() sur un tibble qui contient des données stockées dans Spark, vous verrez une liste contenant un objet de connexion Spark, et quelques autres éléments.

+ Si vous voulez voir un résumé de ce que chaque colonne contient dans le jeu de données auquel le tibble fait référence, vous devez appeler **glimpse()** à la place. 

+ Notez que pour les données distantes telles que celles stockées dans les ensembles de données d'un cluster Spark, le nombre de lignes est un mensonge ! Dans ce cas, glimpse() ne prétend jamais que les données ont plus de 25 lignes.

####**Filtrage des lignes**

+ Outre la sélection des colonnes, l'autre moyen d'extraire les parties importantes de votre ensemble de données consiste à filtrer les lignes. 

+ Pour ce faire, vous utilisez **la fonction filter()**. 

+ Pour utiliser **la fonction filter()**, vous lui passez un tibble et quelques conditions logiques. 

+ Par exemple, pour ne renvoyer que les lignes dont les valeurs de la colonne $x$ sont supérieures à zéro et dont les valeurs de $y$ sont égales à celles de $z$, vous utiliserez la fonction suivante.

      a_tibble %>%
        filter(x > 0, y == z)

+ Avant de vous lancer dans cet exercice, tenez compte de deux avertissements. 

+ Premièrement, ne confondez pas **la fonction filter() de dplyr avec la fonction filter() du paquet stats**. 

+ Deuxièmement, **sparklyr convertit votre code dplyr en code de base de données SQL avant de le transmettre à Spark**. 

+ Cela signifie que seul un nombre limité d'opérations de filtrage est actuellement pris en charge. 

+ Par exemple, vous ne pouvez pas filtrer les lignes de caractères en utilisant des expressions régulières avec un code tel que

      a_tibble %>%
        filter(grepl("a regex", x))

+ ***La page d'aide de translate_sql() décrit les fonctionnalités disponibles***. 

+ ***Vous pouvez utiliser des opérateurs de comparaison comme >, != et %in%, des opérateurs arithmétiques comme +, ^ et %% et des opérateurs logiques comme &, | et !.*** 

+ De nombreuses fonctions mathématiques telles que **log(), abs(), round() et sin()** sont également prises en charge.

*Comme précédemment, l'indexation des crochets ne fonctionne pas actuellement.*

####**Classement des rangées**
+ À l'époque où la musique était stockée sur des CD, un problème récurrent se posait :

  + Par ordre d'artiste ?

+ ***La fonction arrange() vous permet de réorganiser les rangées d'un tibble.*** 

+ Elle prend un tibble, suivi des noms non cités des colonnes.

####**Transformation de variables continues en variables logiques**

+ Les variables logiques sont intéressantes car il est souvent plus facile de penser aux choses en termes de "oui ou non" plutôt qu'en termes numériques.

#####**Transformation de variables continues en variables catégorielles (1)**

+ Une généralisation de l'idée précédente consiste à avoir plusieurs seuils ; c'est-à-dire que vous divisez une variable continue en "buckets" (ou "bins"), tout comme le fait un histogramme. 

+ En base-R, vous utiliserez cut() pour cette tâche. 

+ Par exemple, dans une étude sur les habitudes tabagiques, vous pourriez prendre le nombre typique de cigarettes fumées par jour, et le transformer en un facteur.

        smoking_status <- cut(
          cigarettes_per_day,
          breaks = c(0, 1, 10, 20, Inf),
          labels = c("non", "light", "moderate", "heavy"),
          right  = FALSE
        )

+ **L'équivalent sparklyr de ceci est d'utiliser ft_bucketizer()**. 

+ Le code prend un format similaire à **ft_binarizer()**, mais cette fois vous devez passer un vecteur de points de coupe à l'argument splits. 

+ Voici le même exemple réécrit dans le style sparklyr.

      smoking_data %>%
        ft_bucketizer("cigarettes_per_day", "smoking_status", splits = c(0, 1, 10, 20, Inf))


+ Il y a plusieurs choses importantes à noter. 

+ Vous avez peut-être remarqué que l'argument breaks de **cut()** est le même que l'argument splits de **ft_bucketizer()**. 

+ Il y a une légère différence dans la façon dont les valeurs à la limite sont traitées.

+  Dans **cut()**, par défaut, la limite supérieure (droite) est incluse dans chaque godet, mais pas la gauche. 

+ **ft_bucketizer() inclut la limite inférieure (gauche) dans chaque godet, mais pas la droite. Cela signifie qu'il est équivalent d'appeler cut() avec l'argument right = FALSE.**

+ **Une exception est que ft_bucketizer() inclut les valeurs sur les deux limites pour le seau le plus haut**. 

+ **Ainsi, ft_bucketizer() est également équivalent au paramétrage de include.lowest = TRUE lors de l'utilisation de cut().**

+ La dernière chose à noter est qu'alors que **cut() renvoie un facteur, ft_bucketizer() renvoie un vecteur numérique, avec les valeurs du premier seau renvoyées comme zéro, les valeurs du deuxième seau renvoyées comme un, les valeurs du troisième seau renvoyées comme deux, et ainsi de suite**. 

+ Si vous voulez travailler sur les résultats dans R, vous devez explicitement convertir en facteur. Il s'agit d'un modèle de code courant :

      a_tibble %>%
        ft_bucketizer ("x", "x_buckets", splits = splits) %>%
        collect() %>%
          mutate(x_buckets = factor(x_buckets, labels = labels)

####**Réduire les données par échantillonnage**

+ Lorsque vous travaillez avec un grand ensemble de données, vous n'avez généralement pas besoin de travailler avec toutes les données en même temps. 

+ En particulier au début de votre projet, pendant que vous expérimentez sauvagement ce que vous voulez faire, vous pouvez souvent itérer plus rapidement en travaillant sur un sous-ensemble plus petit des données. 

+ **sdf_sample() fournit un moyen pratique de le faire**. 

+ Elle prend un tibble, et la fraction de lignes à retourner. 

+ Dans ce cas, vous voulez échantillonner sans remplacement. 

+ Pour obtenir un échantillon aléatoire d'un dixième de votre ensemble de données, vous devez utiliser le code suivant.

      a_tibble %>%
        sdf_sample(fraction = 0.1, replacement = FALSE)


+ Étant donné que les résultats de l'échantillonnage sont aléatoires et que vous souhaiterez probablement réutiliser l'ensemble de données rétréci, il est courant d'utiliser **compute()** pour stocker les résultats dans un autre cadre de données Spark.

####**Arbres boostés par le gradient : modélisation**

+ ***Le boosting de gradient est une technique permettant d'améliorer les performances d'autres modèles.***

+ L'idée est d'exécuter un modèle faible mais facile à calculer.

+  Ensuite, vous remplacez les valeurs de la réponse par les résidus de ce modèle, et vous ajustez un autre modèle. 

+ En "additionnant" le modèle de prédiction de la réponse originale et le nouveau modèle de prédiction des résidus, vous obtenez un modèle plus précis. 

+ Vous pouvez répéter ce processus à l'infini, en exécutant de nouveaux modèles pour prédire les résidus des modèles précédents, et en ajoutant les résultats. 

+ À chaque itération, le modèle devient de plus en plus fort.

+ Pour donner un exemple plus concret, ***sparklyr utilise des arbres boostés par gradient, c'est-à-dire des arbres de décision comme modèle faible mais facile à calculer.***

####**Comparaison des performances des modèles**

+ Le tracé vous donne une bonne idée des endroits où le modèle est performant et de ceux où il ne l'est pas. 

+ Parfois, il est intéressant d'avoir une statistique qui vous donne un score pour le modèle. 

+ Ainsi, vous pouvez quantifier la qualité d'un modèle et effectuer des comparaisons entre plusieurs modèles. 

+ Une statistique courante est l'erreur quadratique moyenne (parfois abrégée en "RMSE"), qui élève simplement les résidus au carré, puis prend la moyenne, puis la racine carrée. 

+ Un petit score RMSE pour un ensemble de données donné implique une meilleure prédiction. 

+ (Par défaut, vous ne pouvez pas comparer différents ensembles de données, mais uniquement différents modèles sur le même ensemble de données. Il est parfois possible de normaliser les jeux de données afin de pouvoir les comparer).

*Ici, vous allez comparer les arbres boostés par le gradient et les modèles de forêt aléatoire.*

####**Instructions**

*both_responses, contenant l'année prédite et l'année réelle de la piste à partir des deux modèles, a été prédéfini comme un tibble local.*

+ Créez un ensemble de données de la somme des carrés des résidus.

+ Ajoutez une colonne résiduelle, égale à la réponse prédite moins la réponse réelle.

+ Regroupez les données par modèle.

+ Calculez une statistique sommaire, rmse, égale à la racine carrée de la moyenne des résidus au carré.

```r
# both_responses has been pre-defined
both_responses

both_responses %>%
  # Add a residual column
  mutate(
    residual = predicted - actual
  ) %>%
  # Group by model
  group_by(model) %>%
  # Calculate the root mean square error
  summarize(
    rmse = sqrt(mean(residual ^ 2))
  )
```
