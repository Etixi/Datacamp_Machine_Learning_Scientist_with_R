###**L’astuce du noyau**

####**1. L’astuce du noyau**

+ ***Dans cette leçon, nous ferons nos premiers pas pour comprendre comment les concepts que nous avons appris pour les SVM linéaires peuvent être étendus aux problèmes de classification avec des limites de décision complexes.***

####**2. L’idée de base**

+ ***L’idée de base est de concevoir une transformation mathématique qui rend le problème linéairement séparable.*** 

+ ***Nous verrons comment le faire explicitement pour un ensemble de données radialement séparable, c’est-à-dire qui a une limite circulaire. Nous aborderons les limites plus générales dans le chapitre suivant.***

####**3. L’ensemble de données séparables radialement**
+ Pour rappel, voici un tracé du jeu de données radialement séparable que nous avons créé précédemment. 

+ Les données sont séparées en deux classes, la limite étant un cercle de rayon 0 point-7 unités. 

+ Comment transformeriez-vous les variables $x_1$ et $x_2$ de manière à faire de la frontière une ligne droite au lieu d’un cercle ?

####**4. Transformer le problème**

+ OK, vous pourriez raisonner comme suit: 
  + nous savons que l’équation de la frontière est $x_1$ au carré plus $x_2$ au carré est égal à 0 point 49. 
  
  + Alors, mappons $x_1$ au carré à une nouvelle variable $X_1$ et $x_2$ au carré à une nouvelle variable $X_2$. 
  
  + En ce qui concerne les nouvelles variables, l’équation de la frontière devient $X_1$ plus $X_2$ égale à 0 point-49, qui est une ligne droite. 
  
  + Créons un tracé de la frontière à l’aide des variables transformées.

####**5. Tracer dans l’espace $X_1$-$X_2$ - code**


+ Nous allons utiliser ggplot() pour tracer le jeu de données transformé. 

+ Notez que l’équation de la droite dans l’espace transformé est $X_2$ égale à $-X_1$ plus 0-point-49, donc la pente de la droite est -1 et l’interception est 0-point-49. 

+ Nous pouvons maintenant écrire le code ggplot() comme suit. Voyons à quoi ressemble l’intrigue.

####**6. Tracer dans l’espace $X_1 - X_2$**

+ Comme prévu, la limite de décision est une ligne droite dans l’espace transformé. 

+ Notre transformation a rendu le problème linéairement séparable. 

+ En d’autres termes, si nous choisissons de résoudre le problème en termes de variables originales, alors nous devrions utiliser un quadratique (ou polynôme du deuxième degré) au lieu d’une ligne droite.

####**7. Le noyau polynomial - Partie 1**

+ Le noyau polynomial a la forme générale illustrée dans la diapositive. 

+ Le « degré » est le degré du polynôme. gamma et coef0 sont des paramètres de réglage, et $u$ et $v$ sont deux points de données. 

+ Maintenant, puisque nous savions déjà que notre problème était radialement séparable, nous pouvions deviner que nous avions besoin d’un polynôme de 2e degré, mais nous ne connaissions pas sa forme exacte.

####**8. Fonctions du noyau**

+ Il s’avère qu’en raison de la formulation mathématique des SVM, on ne peut pas choisir n’importe quelle transformation. 

+ La fonction de transformation doit avoir des propriétés spécifiques. Les fonctions qui satisfont à ces propriétés sont appelées fonctions du noyau. 

+ Pour ceux d’entre vous qui connaissent l’algèbre vectorielle, les fonctions du noyau sont des généralisations de produits ponctuels. 

+ Si vous n’êtes pas familier avec l’algèbre vectorielle, rappelez-vous simplement que l’idée de base est de trouver une fonction de noyau qui sépare bien les données.

####**7. Le modèle réglé**

+ OK, nous avons donc construit le modèle réglé en utilisant les meilleures valeurs de coût et de gamma. 

+ La précision du test s’avère être de 95%, ce qui est légèrement meilleur que celui du cas par défaut, mais regardons le graphique.

####**8. Limite de décision réglée**

+ OK, donc l’intrigue montre clairement que le modèle réglé fait un bien meilleur travail pour capturer le chiffre réel de la limite 8. 

+ Cela indique que le modèle réglé est en effet considérablement supérieur au modèle par défaut. 

+ Dans ce cas, nous pourrions utiliser cette méthode de vérification parce que nous connaissions la limite réelle de la décision. 

+ Dans les situations réelles, ce ne sera pas le cas. 

+ Ce qui reste vrai en général, c’est que la nature locale du noyau RBF lui permet de capturer les nuances d’une frontière complexe, ce qui n’est tout simplement pas possible via un noyau linéaire ou polynomial.
