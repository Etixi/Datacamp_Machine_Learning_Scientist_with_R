####**4. La fonction de carte**
+ **La fonction map() applique une fonction souhaitée à chaque élément d’un vecteur ou d’une liste et retourne toujours une liste comme résultat**.

+ Cette fonction nécessite deux paramètres, point $x$ et point $f$.

####**5. La fonction de carte**
+ Le point $x$ est le vecteur ou la liste sur lequel vous souhaitez itérer tandis que le point $f$ est la fonction.

+ La fonction peut être une fonction prédéfinie ou anonyme utilisant la syntaxe de la formule.

####**6. La fonction de carte**

+ Par exemple, si vous souhaitez utiliser la **fonction mean()**, vous pouvez vous y référer directement ou vous pouvez créer une fonction anonyme en utilisant le tilda pour indiquer que vous utilisez une formule et le point $x$ pour indiquer l’espace réservé de valeur.

####**7. 3) Métrique: Précision**

+ La prochaine mesure que nous examinerons est la précision.

+ Cette métrique calcule la fréquence à laquelle le modèle est correct pour prédire la classe TRUE.

+ Vous le calculez à l’aide de la fonction precision().

####**8. 3) Mesure : Rappel**

+ Enfin, il y a le rappel. Cette mesure compare le nombre d’observations correctement identifiées par le modèle comme TRUE au nombre total d’observations TRUE.

+ En d’autres termes, il mesure la vitesse à laquelle le modèle peut capturer la classe TRUE.

+ Si vous souhaitez créer un modèle qui capturerait autant d’employés risqués que possible, vous devriez envisager cette mesure.
