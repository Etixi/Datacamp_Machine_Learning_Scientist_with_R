####**2. Apprentissage supervisé**

+ **Aujourd’hui, caret est l’un des progiciels les plus utilisés dans R pour l’apprentissage supervisé (également connu sous le nom de modélisation prédictive)**. 

+ **L’apprentissage supervisé est l’apprentissage automatique lorsque vous avez une « variable cible » ou quelque chose de spécifique que vous souhaitez prédire**. 

+ Un exemple classique d’apprentissage supervisé consiste à prédire de quelle espèce est un iris, en fonction de ses mesures physiques. 

+ Un autre exemple serait de prédire quels clients de votre entreprise vont « se désabonner » ou annuler leur service. 

+ Dans ces deux cas, nous avons quelque chose de spécifique que nous voulons prédire à partir de nouvelles données : l’espèce et le barat.

####**3. Apprentissage supervisé**

+ Il existe deux principaux types de modèles prédictifs : 
  + **la classification et la régression**. 
  
+ **Les modèles de classification prédisent des variables qualitatives, par exemple l’espèce d’une fleur, ou « un client va-t-il se désabonner »**. 

+ **Les modèles de régression prédisent des variables quantitatives, par exemple le prix d’un diamant**. 

+ Une fois que nous avons un modèle, **nous utilisons une « métrique » pour évaluer le fonctionnement du modèle**. 

+ **Une métrique est quantifiable et nous donne une mesure objective de la façon dont le modèle prédit sur de nouvelles données**. 

+ Pour les problèmes de régression, nous nous concentrerons sur **« l’erreur quadratique moyenne » ou $RMSE$ comme mesure de choix**. 

+ **Il s’agit de l’erreur que les modèles de régression linéaire cherchent généralement à minimiser**, par exemple dans la fonction **lm() de R**. 

+ *C’est une bonne mesure d’erreur à usage général, et la plus courante pour les modèles de régression*.

####**4. Évaluation de la performance du modèle**

+ Malheureusement, il est courant de calculer RMSE sur les mêmes données que nous avons utilisées pour ajuster le modèle. 

+ Cela conduit généralement à des estimations trop optimistes des performances des modèles. 

+ *C’est ce qu’on appelle aussi le surajustement*. 
+ ***Une meilleure approche consiste à utiliser des estimations hors échantillon de la performance du modèle***. 

+ ***C’est l’approche que prend Caret, car elle simule ce qui se passe dans le monde réel et nous aide à éviter le sur-ajustement***.

####**5. Erreur dans l’échantillon**

+ Cependant, il est utile de commencer par examiner *l’erreur dans l’échantillon, afin que nous puissions la comparer plus tard avec l’erreur hors échantillon sur le même ensemble de données*. 

+ Tout d’abord, nous chargeons le jeu de données mtcars et ajustons un modèle aux 20 premières lignes. 

+ Ensuite, nous faisons des prédictions dans l’échantillon, en utilisant la fonction de prédiction de notre modèle. 

+ Enfin, nous calculons $RMSE$ sur nos données d’entraînement et obtenons d’assez bons résultats.

###**Mesures d’erreur hors échantillon**

####**1. Mesures d’erreur hors échantillon**

+ Salut! Je suis Zach Deane Mayer, et je suis l’un des co-auteurs du paquet caret. 

+ J’ai une passion pour la science des données et je passe la plupart de mon temps à travailler et à réfléchir aux problèmes de l’apprentissage automatique.

####**2. Erreur hors échantillon**

+ Ce cours se concentre sur **la modélisation prédictive plutôt qu’explicative**. 

+ **Nous voulons des modèles qui ne correspondent pas trop aux données d’entraînement et qui généralisent bien**. 

+ En d’autres termes, notre principale préoccupation lors de la modélisation est la suivante : **« Les modèles fonctionnent-ils bien sur les nouvelles données ? »** 

+ La meilleure façon de répondre à cette question est de **tester les modèles sur de nouvelles données**. 

+ Cela simule l’expérience du monde réel, dans laquelle vous vous adaptez à un ensemble de données, puis prédites sur de nouvelles données, où vous ne connaissez pas réellement le résultat. 

+ **Simuler cette expérience avec une séparation train/test vous aide à faire une évaluation honnête de vous-même en tant que modéliste**. 

+ **C’est l’un des enseignements clés de l’apprentissage automatique** : 
  + **les mesures d’erreur** doivent être calculées sur de **nouvelles données**, 
  + car la validation dans l’échantillon (ou la prédiction sur vos données d’entraînement) garantit essentiellement **un surajustement**. 
  
+ La validation hors échantillon vous aide à choisir des modèles qui continueront à bien fonctionner à l’avenir. 
  
+ C’est l’objectif principal du **package caret** en général et de ce cours en particulier: ne pas trop s’adapter. 

+ Choisissez des modèles qui fonctionnent bien sur de nouvelles données.

####**3. Exemple : RMSE hors échantillon**


+ Passons en revue un exemple simple de validation hors échantillon : 
  + nous commençons par un modèle de régression linéaire, adapté aux 20 premières lignes de l’ensemble de données **mtcars**. 
  
  + Ensuite, nous faisons des prédictions avec ce modèle sur un NOUVEL ensemble de données: 
    + les 12 dernières observations de l’ensemble de données mtcars. 
    
+ Les 12 voitures de cet ensemble de tests ne seront pas utilisées pour déterminer les coefficients du modèle de régression linéaire, et constituent donc un bon test de la façon dont nous pouvons prédire sur de nouvelles données. 

+ En pratique, plutôt que de fractionner manuellement le jeu de données, nous utiliserions la fonction **createResamples ou createFolds dans caret, mais la division manuelle simplifie cet exemple**. 

+ Enfin, nous calculons **l’erreur quadratique moyenne racine (ou RMSE) sur l’ensemble de test en comparant les prédictions de notre modèle aux valeurs MPG réelles pour l’ensemble de test**. 

+ **RMSE est une mesure de l’erreur moyenne du modèle.** 

+ Il a les mêmes unités que l’ensemble de test, ce qui signifie que notre modèle est décalé de 5 à 6 miles par gallon, en moyenne.

####**4. Comparer avec RMSE dans l’échantillon**

+ Par rapport au RMSE dans l’échantillon à partir d’un modèle adapté à l’ensemble de données complet, notre modèle est nettement pire. 

+ Si nous avions utilisé l’erreur dans l’échantillon, nous nous serions trompés en pensant que notre modèle est bien meilleur qu’il ne l’est en réalité. 

+ Il est difficile de faire des prédictions sur de nouvelles données, comme le montre cet exemple. 

+ L’erreur hors échantillon aide à tenir compte de ce fait, ce qui nous permet de nous concentrer sur les modèles qui prédisent des choses que nous ne savons pas déjà.

###**Validation croisée**

####**1. Validation croisée**
+ Dans la dernière partie, **nous avons divisé manuellement nos données**

####**2. Validation croisée**
+ **Nous avons divisé manuellement nos données en un seul ensemble de tests et évalué une fois l’erreur hors échantillon**.

####**3. Validation croisée**

+ Cependant, ce processus est un peu fragile : 

  + *la présence ou l’absence d’une seule valeur aberrante peut modifier considérablement notre RMSE hors échantillon.*

####**4. Validation croisée**

+ *Une meilleure approche qu’une simple séparation train/test consiste à utiliser plusieurs ensembles de tests et à faire la moyenne des erreurs hors échantillon, ce qui nous donne une estimation plus précise de la véritable erreur hors échantillon*. 

+ *L’une des approches les plus courantes pour plusieurs ensembles de tests est connue sous le nom de « validation croisée », dans laquelle nous divisons nos données en dix « plis » ou fractionnements train/test*. 

+ Nous créons ces plis de telle sorte que chaque point de notre jeu de données se trouve dans exactement un ensemble de tests. 

+ Cela nous donne 10 ensembles de tests, et mieux encore, signifie que chaque point de notre ensemble de données se produit exactement une fois. 

+ En d’autres termes, nous obtenons un ensemble de tests de la même taille que notre ensemble d’entraînement, mais composé de prédictions hors échantillon!

####**5. Validation croisée**


+ Nous assignons chaque ligne à son ensemble de test unique de manière aléatoire, afin d’éviter tout type de biais systémique dans nos données. 

+ C’est l’un des meilleurs moyens d’estimer l’erreur hors échantillon pour les modèles prédictifs.

###**Matrice de confusion**

####**1. Matrice de confusion**


+ **Un outil très utile pour évaluer les modèles de classification binaire est connu sous le nom de « matrice de confusion »**. 

+ **Il s’agit d’une matrice des classes prédites du modèle par rapport aux résultats réels dans la réalité**. 

+ **C’est ce qu’on appelle une matrice de confusion parce qu’elle révèle à quel point le modèle est « confus » entre les 2 classes et met en évidence les cas dans lesquels une classe est confondue avec l’autre**.

####**2. Matrice de confusion**

+ *Les colonnes de la matrice de confusion sont les classes vraies, tandis que les lignes de la matrice de confusion sont les classes prédites*. 

+ De gauche à droite, de haut en bas, les cellules de la matrice sont : 
  + les vrais positifs, 
  + les faux positifs, 
  + les faux négatifs et les vrais négatifs. 
  
+ **La diagonale principale de la matrice de confusion est les cas où le modèle est correct (vrais positifs et vrais négatifs) et la deuxième diagonale de la matrice de confusion est les cas où le modèle est incorrect (faux négatifs et faux positifs)**. 

+ Passons brièvement en revue les 4 résultats possibles avec un modèle de classification binaire: **Les vrais positifs sont des cas où le modèle a correctement prédit oui**. 

+ **Les faux positifs sont des cas où le modèle a prédit incorrectement oui**. 

+ **Les faux négatifs sont des cas où le modèle a prédit à tort non**. 

+ **Et les vrais négatifs sont des cas où le modèle a correctement prédit non**. 


+ Ces 4 résultats sont importants lors de **l’évaluation de la précision d’un modèle prédictif**, il est donc utile de les examiner simultanément dans un seul tableau.

####**3. Matrice de confusion**
+ Pour générer une matrice de confusion, nous commençons par ajuster un modèle à notre ensemble d’entraînement.

+ Dans le cas le plus simple, nous utilisons une probabilité de 50% comme seuil et attribuons tout ce qui est inférieur à 50% comme roche et tout ce qui dépasse 50% comme mine.

###**Présentation de la courbe ROC**

####**1. Présentation de la courbe ROC**
+ **L’évaluation manuelle des seuils de classification est un travail difficile**!

####**2. Le défi**


+ Pour le faire correctement, nous devrions calculer manuellement des dizaines (ou des centaines) de matrices de confusion, puis les inspecter visuellement jusqu’à ce que nous en trouvions une que nous aimons. 

+ Cela semble non scientifique, car cela nécessite beaucoup de travail manuel, est basé sur l’heuristique et pourrait facilement négliger un seuil important particulier. 

+ Nous avons besoin d’une approche plus systématique pour évaluer les seuils de classification.

####**3. Courbes ROC**


+ ***Une approche courante de ce problème consiste à laisser l’ordinateur évaluer de manière itérative tous les seuils de classification possibles, puis à calculer le taux de vrais positifs et le taux de faux positifs pour chacun d’eux***. 

+ ***Nous pouvons ensuite tracer le taux réel postif / faux positif à chaque seuil possible, et visualiser le compromis entre les 2 modèles extrêmes (prédire toutes les mines vs prédire toutes les roches, ou 100% vrai taux positif vs 0% taux de faux positifs)***.

+ ***La courbe résultante est appelée courbe ROC, ou courbe caractéristique de fonctionnement du récepteur***. (Ne vous inquiétez pas, personne ne se souvient vraiment de cet acronyme.) 

+ ***La courbe ROC a été développée pendant la 2e guerre mondiale comme méthode d’analyse des signaux radar***. 

+ ***Dans ce cas historiquement intéressant, un vrai positif serait d’identifier correctement un bombardier par son signal radar, tandis qu’un faux positif identifierait une volée d’oiseaux comme un bombardier***.

###**Surface sous la courbe (ASC)- (ROC)**

####**1. Surface sous la courbe (ASC)- (ROC)**
+ Le simple fait de regarder une courbe ROC commence à nous donner une bonne idée de la façon d’évaluer si notre modèle prédictif est bon ou non.

####**2. De ROC à AUC**

+ **Une observation intéressante est que les modèles avec des prédictions aléatoires ont tendance à produire des courbes qui suivent de près la ligne diagonale**. 

+ D’autre part, **les modèles avec un seuil de classification qui permet une séparation parfaite des classes produisent une « boîte » avec un seul point à (1,0) pour représenter un modèle où il est possible d’atteindre un taux de vrais positifs de 100% et un taux de faux positifs de 0%**. 

+ Ne serait-ce pas bien? 

+ En continuant avec cet exemple, si nous calculons l’aire sous chacune de ces 2 courbes ROC, une propriété intéressante émerge: **l’aire sous la courbe pour un modèle parfait est exactement 1, car notre graphique représente un carré de 1 par 1, et l’aire moyenne sous la courbe pour un modèle aléatoire est le point 5, car notre graphique représente une ligne diagonale**.

####**3. Définition de l' AUC**

+ ***Nous pouvons utiliser ces informations pour formaliser une mesure de la précision du modèle appelée « AUC » ou « zone sous la courbe »***. 

+ ***Cette métrique est calculée sur la base du tracé de la courbe ROC et est extrêmement utile***. 

+ Il s’agit d’un résumé numérique de la précision du modèle qui ne nous oblige pas à évaluer manuellement les matrices de confusion. 

+ Ce nombre résume les performances du modèle sur tous les seuils de classification possibles et constitue une mesure unique que nous pouvons utiliser pour classer différents modèles dans le même ensemble de données.

####**4. Définition de l' AUC**


+ **Il varie de 0 à 1, où le point-5 est le ROC d’un modèle aléatoire et 1-point-0 est le ROC d’un modèle parfait**. 

+ (Un modèle parfaitement anti-prédictif aurait une ROC de 0, mais cela arrive rarement). 

+ En pratique, la plupart des modèles se situent entre le point 5 et le point 0, tandis qu’un très mauvais modèle peut parfois être dans la gamme point-4. 

+ En règle générale, l’AUC peut être considérée comme une lettre, où le point-9 est un « A », le point 8 est un « B », le point 7 est un « C », le point 5 est un « F », et ainsi de suite. Je suis généralement satisfait d’un modèle qui a une ASC de point-8 ou plus, et les modèles dans la gamme point-7 sont souvent utiles.

###**Présentation de glmnet**

####**1. Présentation de glmnet**
+ Nous allons maintenant présenter l’un de mes modèles prédictifs préférés : **le modèle glmnet**.

####**2. Présentation de glmnet**

+ ***Les modèles Glmnet sont une extension des modèles linéaires généralisés (ou la fonction glm dans R)***. 

+ **Cependant, ils ont une sélection de variables intégrée qui est utile sur de nombreux ensembles de données du monde réel**. 

+ **En particulier, cela aide les modèles de régression linéaire à mieux gérer la colinéarité - ou la corrélation entre les prédicteurs d’un modèle - et les empêche également d’être trop confiants dans les résultats dérivés de petites tailles d’échantillons**. 

+ Il existe 2 formes principales de modèles glmnet : 
  + **la régression lasso, qui pénalise le nombre de coefficients non nuls**, 
  + **et la régression de crêtes, qui pénalise l’amplétude absolue des coefficients**. 
  
+ Ces pénalités sont calculées lors de l’ajustement du modèle et sont utilisées par l’optimiseur pour ajuster les coefficients de régression linéaire. 

+ En d’autres termes, **un glmnet tente de trouver un modèle parcimonieux, avec soit quelques coefficients non nuls, soit de petits coefficients de magnitude absolue, qui correspondent le mieux à l’ensemble de données d’entrée**. 

+ *Il s’agit d’un modèle extrêmement utile qui se marie particulièrement bien avec les modèles forestiers aléatoires, car il a tendance à donner des résultats différents*.

####**3. Réglage des modèles glmnet**


+ Les modèles GLMNET sont une combinaison de 2 types de modèles : 
  + **la régression LASSO (avec une pénalité sur le nombre de coefficients non nuls)** 
  + et **la régression de crête (avec une pénalité sur les grands coefficients)**. 
  
+ De plus, les modèles glmnet peuvent s’adapter à un * mélange * de **modèles lasso et ridge, c’est-à-dire un modèle avec une petite pénalité à la fois sur le nombre de coefficients non nuls et leur magnitude absolue**. 

+ Cela donne aux modèles glmnet de nombreux paramètres à régler. 

+ Le paramètre alpha est compris entre 0 et 1, où 0 est une régression pure de la crête, 1 est une régression lasso pure et toute valeur entre est un mélange des deux. 

+ Lambda, d’autre part, varie de 0 à l’infini positif et contrôle la taille de la pénalité. 

+ Des valeurs plus élevées de lambda donneront des modèles plus simples, et des valeurs suffisamment élevées de lambda donneront des modèles d’interception uniquement qui prédisent simplement la moyenne de la variable de réponse dans les données d’apprentissage.

###**Imputation médiane**

####**1. Imputation médiane**
+ Les données du monde réel ont des valeurs manquantes.

####**2. Traitement des valeurs manquantes**

+ C’est un problème pour la plupart des algorithmes statistiques ou d’apprentissage automatique : 
  + ils nécessitent généralement des chiffres pour travailler avec, 
  + et ne savent pas quoi faire avec les données manquantes. 
  
+ Une approche courante consiste à jeter les lignes avec des données manquantes, mais ce n’est généralement pas une bonne idée. 

+ Cela peut entraîner des biais dans votre ensemble de données et générer des modèles trop confiants. 

+ Cela peut aussi, dans des cas extrêmes, vous amener à jeter toutes vos données. 

+ Une bien meilleure stratégie consiste à **utiliser la médiane pour deviner ce que serait une valeur manquante, si elle n’était pas manquante**. 

+ C’est une très bonne idée si vos données sont « manquantes au hasard » et vous permet de modéliser des données qui incluent des lignes avec des valeurs manquantes.

####**3. Exemple : mtcars**
+ Générons des données avec des valeurs manquantes. 

+ Nous commencerons par l’ensemble de données mtcars, qui contient des mesures des caractéristiques physiques de certaines voitures. 

+ Dans ce cas, nous voulons prédire le MPG de la voiture, en fonction des autres attributs de la voiture. 

+ Supposons que certains constructeurs ne déclarent pas la puissance de leur voiture et remplaçons au hasard certains points de cette colonne par des valeurs manquantes. 

+ Nous pouvons ensuite diviser l’ensemble de données en un cadre de données de prédicteurs ($X$) et la cible que nous voulons prédire ($Y$). 

+ Cela démontre l’interface non-formule de caret pour la modélisation. 

+ Malheureusement, en raison des valeurs manquantes dans $X$, lorsque nous allons ajuster le modèle, il échoue avec une erreur cryptique. 

+ C’est un point où de nombreux nouveaux utilisateurs sont bloqués et doivent venir chercher de l’aide.

####**4. Une solution simple**

+ La solution simple à ce problème est de passer **« medianImpute »** à l’argument preProcess pour train, qui indique à caret d’imputer les valeurs manquantes dans $X$ avec leurs médianes. 

+ caret fait en fait cette imputation à l’intérieur de chaque pli de la validation croisée, de sorte que vous obtenez une évaluation honnête de l’ensemble du processus de modélisation: le modèle de forêt aléatoire est adapté après l’imputation. 

+ Ce modèle s’exécute désormais sans erreur et ne nécessite plus que vous, en tant que data scientist, effectuiez un travail supplémentaire pour nettoyer vos données.

###**Imputation KNN**

####**1. Imputation KNN**
+ **L’imputation médiane** pose certains problèmes.

####**2. Traitement des valeurs manquantes**

+ C’est très rapide, mais cela peut produire des **résultats incorrects si les données d’entrée ont un biais systématique et manquent non pas au hasard**. 

+ En d’autres termes, **s’il y a une tendance dans les données qui mène à des valeurs manquantes, l’imputation médiane peut ne pas le faire**.

+ **Un type utile d’imputation de valeur manquante est k-plus proche-voisins, ou imputation knn**. 

+ Il s’agit d’une stratégie d’imputation des valeurs manquantes basée sur d’autres lignes non manquantes « similaires ». 

+ **Cette méthode tente de surmonter le problème des manquants non aléatoires en déduisant quelle serait la valeur manquante, sur la base d’observations similaires dans d’autres variables non manquantes**.

####**3. Exemple : manquant pas au hasard**

###**Analyse des composantes principales (ACP)**

####**1. Analyse des composantes principales (ACP)**


+ **L’analyse des composants principaux (ou ACP) est l’une de mes étapes de prétraitement préférées pour les modèles de régression linéaire**. 

+ Vous remarquerez que je l’ai utilisé comme exemple dans de nombreuses vidéos précédentes.

####**2. Analyse des composantes principales**

+ ***L’ACP est incroyablement utile car elle combine toutes les variables à faible variance et corrélées de votre ensemble de données en un seul ensemble de prédicteurs perpendiculaires à forte variance***. 

+ Comme nous l’avons vu précédemment, les variables de faible variance peuvent être problématiques pour la validation croisée, mais peuvent également contenir des informations utiles. 

+ Il est préférable de trouver un moyen systématique d’utiliser cette information, plutôt que de la jeter. 

+ De plus, les prédicteurs perpendiculaires sont utiles car ils ne sont parfaitement pas corrélés. 

+ ***Les modèles de régression linéaire ont du mal avec la corrélation entre les variables (également connue sous le nom de colinéarité), et l’ACP élimine élégamment ce problème de l’équation***.

####**3. ACP : une représentation visuelle**


+ ***L’ACP recherche des combinaisons linéaires à variance élevée des données d’entrée qui sont perpendiculaires les unes aux autres***. 

+ ***La première composante de l’ACP est la composante de variance la plus élevée et l’axe de variance la plus élevée de l’ensemble de données d’origine***. 

+ ***La deuxième composante de l’ACP a la deuxième variance la plus élevée, et ainsi de suite***. 

+ Ce diagramme illustre le fonctionnement de l’ACP. 

+ Nous avons 2 variables corrélées, $x$ et $y$. 

+ Lorsqu’ils sont tracés ensemble, nous pouvons voir leur relation. 

+ L’ACP transforme les données par rapport à cette corrélation et trouve une nouvelle variable (la longue flèche diagonale pointant vers le haut et vers la droite) qui reflète la corrélation partagée de $x$ et $y$. 

+ Après avoir trouvé la première composante ACP, la deuxième composante ACP est contrainte d’être perpendiculaire, et est la deuxième flèche allant vers le haut et vers la gauche. 

+ En d’autres termes, la première composante de l’ACP reflète la similitude entre $x$ et $y$, tandis que la deuxième composante de l’ACP met l’accent sur la différence entre $x$ et $y$. 

+ Cette idée est facile à illustrer en 2 dimensions, mais s’étend également à plusieurs dimensions.
